```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import re

# ============================================================
# CONFIGURAÇÃO GERAL DO ADC
# ============================================================

V_MIN = 0.1
V_MAX = 0.7

N_BITS = 4
N_CODES = 2**N_BITS
N_COMPARATORS = N_CODES - 1

# Output digital threshold
VOUT_THRESHOLD = 0.4

# Hold fixo
HOLD_TIME = 20e-9

# Rampas esperadas
ALLOWED_RAMPS_NS = [35, 30, 25, 10, 5]

# LSB ideal
LSB = (V_MAX - V_MIN) / N_CODES

# Se o pulso começa em t=0, deixa None.
# Se houver atraso antes do pulso, mete manualmente, ex: 5e-9
START_TIME_MANUAL = None

# ============================================================
# CRITÉRIOS PASS/FAIL
# ============================================================

MAX_ABS_DNL_LIMIT = 0.5
MAX_ABS_INL_LIMIT = 0.5
MAX_HYST_LIMIT = 1.0
MAX_OFFSET_LIMIT = 0.5

# Se True, o PASS/FAIL usa DNL interno, ignorando code 0 e code 15.
# Isto é útil quando os endpoints dependem muito de V_MIN/V_MAX.
USE_INTERNAL_DNL_FOR_PASS = True

print("===================================")
print("CONFIGURAÇÃO ADC")
print("===================================")
print(f"V_MIN                        = {V_MIN:.3f} V")
print(f"V_MAX                        = {V_MAX:.3f} V")
print(f"N_BITS                       = {N_BITS}")
print(f"N_CODES                      = {N_CODES}")
print(f"N_COMPARATORS                = {N_COMPARATORS}")
print(f"LSB                          = {LSB*1e3:.3f} mV")
print(f"HOLD fixo                    = {HOLD_TIME*1e9:.3f} ns")
print(f"Rampas esperadas             = {ALLOWED_RAMPS_NS} ns")
print(f"VOUT_THRESHOLD               = {VOUT_THRESHOLD:.3f}")
print(f"USE_INTERNAL_DNL_FOR_PASS    = {USE_INTERNAL_DNL_FOR_PASS}")

print()
print("===================================")
print("CRITÉRIOS PASS/FAIL")
print("===================================")
print(f"Max |DNL|          <= {MAX_ABS_DNL_LIMIT:.3f} LSB")
print(f"Max |INL|          <= {MAX_ABS_INL_LIMIT:.3f} LSB")
print(f"Max |Hysteresis|   <= {MAX_HYST_LIMIT:.3f} LSB")
print(f"Max |Offset|       <= {MAX_OFFSET_LIMIT:.3f} LSB")


# ============================================================
# FUNÇÕES AUXILIARES
# ============================================================

def extract_ramp_time_from_filename(filename):
    """
    Extrai o valor da rampa a partir do nome do ficheiro.

    Exemplos aceites:
    adc_ramp_35ns.txt
    adc_ramp_30ns.txt
    adc_ramp_25ns.txt
    adc_ramp_10ns.txt
    adc_ramp_5ns.txt
    """

    name = filename.lower()

    match = re.search(r"ramp[_\-\s]?([0-9]*\.?[0-9]+)\s*ns", name)

    if match is None:
        raise ValueError(
            f"Não consegui extrair ramp time do ficheiro: {filename}\n"
            "Usa nomes tipo: adc_ramp_35ns.txt, adc_ramp_30ns.txt, adc_ramp_25ns.txt"
        )

    ramp_ns = float(match.group(1))

    if ramp_ns not in ALLOWED_RAMPS_NS:
        print()
        print("AVISO:")
        print(f"O ficheiro {filename} tem ramp = {ramp_ns} ns.")
        print(f"As rampas esperadas são {ALLOWED_RAMPS_NS} ns.")
        print("Vou analisar na mesma.")

    ramp_s = ramp_ns * 1e-9

    return ramp_s, ramp_ns


def interp_crossing_vin(t, vin, y, threshold, t0, t1, crossing_type="any"):
    """
    Encontra Vin quando o output digital y cruza threshold.

    crossing_type:
      rising  -> 0 para 1
      falling -> 1 para 0
      any     -> qualquer cruzamento
    """

    mask = (t >= t0) & (t <= t1)

    if np.sum(mask) < 2:
        return np.nan

    vv = vin[mask]
    yy = y[mask]

    d = yy - threshold

    candidates = []

    for i in range(len(d) - 1):

        d0 = d[i]
        d1 = d[i + 1]

        if d0 == 0:
            candidates.append(i)
            continue

        if d0 * d1 > 0:
            continue

        if crossing_type == "rising":
            if not (yy[i] < threshold and yy[i + 1] >= threshold):
                continue

        if crossing_type == "falling":
            if not (yy[i] > threshold and yy[i + 1] <= threshold):
                continue

        candidates.append(i)

    if len(candidates) == 0:
        return np.nan

    i = candidates[0]

    y0, y1 = yy[i], yy[i + 1]
    v0, v1 = vv[i], vv[i + 1]

    if y1 == y0:
        return v0

    alpha = (threshold - y0) / (y1 - y0)
    vin_cross = v0 + alpha * (v1 - v0)

    return vin_cross


def compute_dnl_inl(thresholds_sorted):
    """
    Calcula DNL e INL a partir dos 15 thresholds médios.
    """

    thresholds_sorted = np.asarray(thresholds_sorted)
    thresholds_sorted = thresholds_sorted[np.isfinite(thresholds_sorted)]
    thresholds_sorted = np.sort(thresholds_sorted)

    if len(thresholds_sorted) != N_COMPARATORS:
        return None, None, None, None

    # Edges: V_MIN, th1, th2, ..., th15, V_MAX
    edges = np.concatenate(([V_MIN], thresholds_sorted, [V_MAX]))

    # Widths dos 16 códigos
    widths = np.diff(edges)

    # DNL dos 16 códigos
    dnl = widths / LSB - 1

    # Thresholds ideais
    ideal_thresholds = V_MIN + np.arange(1, N_COMPARATORS + 1) * LSB

    # INL das 15 transições
    inl = (thresholds_sorted - ideal_thresholds) / LSB

    return dnl, inl, ideal_thresholds, widths


def read_waveview_file(filename):
    """
    Lê ficheiro WaveView.

    Usa APENAS:
        v(/out<0>) ... v(/out<14>)

    Ignora:
        v(out<0>) ... v(out<14>)
    """

    header = None
    header_line_idx = None

    with open(filename, "r") as f:
        for i, line in enumerate(f):
            s = line.strip()

            if s.startswith("XVAL"):
                header = s.split()
                header_line_idx = i
                break

    if header is None:
        raise ValueError("Header XVAL não encontrado.")

    out_dict = {}
    ignored_analog = []
    duplicate_out_info = []
    vin_col = None

    for i, name in enumerate(header):

        clean = (
            str(name)
            .replace("&lt;", "<")
            .replace("&gt;", ">")
            .strip()
            .lower()
        )

        # Vin
        if "v(vin" in clean:
            vin_col = i

        # Outputs digitais v(/out<k>)
        m_digital = re.search(r"v\(/out<(\d+)>\)", clean)

        if m_digital:
            idx = int(m_digital.group(1))

            if 0 <= idx <= 14:
                if idx not in out_dict:
                    out_dict[idx] = (i, idx, name)
                else:
                    duplicate_out_info.append((i, idx, name))

            continue

        # Outputs analógicos v(out<k>) ignorados
        m_analog = re.search(r"v\(out<(\d+)>\)", clean)

        if m_analog:
            ignored_analog.append((i, int(m_analog.group(1)), name))

    if vin_col is None:
        raise ValueError("Coluna Vin não encontrada.")

    if len(out_dict) == 0:
        raise ValueError("Nenhuma coluna digital v(/out<0>)...v(/out<14>) encontrada.")

    # Ordenar por out<0>, out<1>, ..., out<14>
    out_info_sorted = [out_dict[k] for k in sorted(out_dict.keys())]

    out_cols = [x[0] for x in out_info_sorted]
    out_indices = [x[1] for x in out_info_sorted]

    print()
    print("------------- COLUNAS USADAS -------------")
    print(f"Vin column                     = {vin_col}")
    print(f"Outputs digitais usados        = {out_indices}")
    print(f"Colunas digitais usadas        = {out_cols}")
    print(f"N outputs digitais usados      = {len(out_indices)}")

    if len(ignored_analog) > 0:
        ignored_indices = [x[1] for x in ignored_analog]
        print(f"Outputs analógicos ignorados   = {ignored_indices}")

    if len(duplicate_out_info) > 0:
        duplicate_indices = [x[1] for x in duplicate_out_info]
        print(f"Outputs digitais duplicados ignorados = {duplicate_indices}")

    if len(out_indices) != N_COMPARATORS:
        print()
        print("AVISO:")
        print(f"Foram usados {len(out_indices)} outputs digitais, mas eram esperados {N_COMPARATORS}.")
        print("Confirma se tens v(/out<0>) até v(/out<14>) no ficheiro.")

    # Ler todas as colunas para evitar problemas de ordem do usecols
    df = pd.read_csv(
        filename,
        sep=r"\s+",
        comment="#",
        skiprows=header_line_idx + 1,
        header=None,
        engine="c"
    )

    data = df.to_numpy(dtype=float)

    t = data[:, 0]
    vin = data[:, vin_col]
    outs = data[:, out_cols]

    # Ordenar por tempo
    idx_sort = np.argsort(t)

    t = t[idx_sort]
    vin = vin[idx_sort]
    outs = outs[idx_sort, :]

    # Remover tempos duplicados
    t_unique, unique_idx = np.unique(t, return_index=True)

    t = t_unique
    vin = vin[unique_idx]
    outs = outs[unique_idx, :]

    return t, vin, outs, out_indices


def analyze_file(filename, make_plots=True):

    # ========================================================
    # TEMPOS DO PULSO
    # ========================================================

    RAMP_TIME, RAMP_TIME_NS = extract_ramp_time_from_filename(filename)

    RISE_TIME = RAMP_TIME
    FALL_TIME = RAMP_TIME
    HOLD_TIME_LOCAL = HOLD_TIME

    PERIOD = RISE_TIME + HOLD_TIME_LOCAL + FALL_TIME

    # Ignorar hold para fs ativo
    T_ACTIVE = RISE_TIME + FALL_TIME
    FS_ACTIVE = 1 / T_ACTIVE

    # Incluir hold para fs sistema
    FS_SYSTEM = 1 / PERIOD

    # ========================================================
    # LER FICHEIRO
    # ========================================================

    t, vin, outs, out_indices = read_waveview_file(filename)

    n_outputs = outs.shape[1]

    print()
    print("===================================")
    print(f"ANÁLISE: {filename}")
    print("===================================")
    print(f"N pontos                  = {len(t)}")
    print(f"N outputs usados          = {n_outputs}")
    print(f"Tempo inicial             = {t[0]*1e9:.3f} ns")
    print(f"Tempo final               = {t[-1]*1e9:.3f} ns")

    print()
    print("------------- CONFIGURAÇÃO DO PULSO -------------")
    print(f"Ramp subida               = {RISE_TIME*1e9:.3f} ns")
    print(f"Hold                      = {HOLD_TIME_LOCAL*1e9:.3f} ns")
    print(f"Ramp descida              = {FALL_TIME*1e9:.3f} ns")
    print(f"Período total waveform    = {PERIOD*1e9:.3f} ns")
    print(f"fs ativo, hold ignorado   = {FS_ACTIVE/1e6:.3f} MHz")
    print(f"fs sistema, com hold      = {FS_SYSTEM/1e6:.3f} MHz")

    # ========================================================
    # JANELAS DO ÚNICO PULSO
    # ========================================================

    if START_TIME_MANUAL is None:
        pulse_start = t[0]
    else:
        pulse_start = START_TIME_MANUAL

    up_t0 = pulse_start
    up_t1 = pulse_start + RISE_TIME

    hold_t0 = up_t1
    hold_t1 = hold_t0 + HOLD_TIME_LOCAL

    down_t0 = hold_t1
    down_t1 = down_t0 + FALL_TIME

    print()
    print("------------- JANELAS USADAS -------------")
    print(f"Subida  : {up_t0*1e9:.3f} ns -> {up_t1*1e9:.3f} ns")
    print(f"Hold    : {hold_t0*1e9:.3f} ns -> {hold_t1*1e9:.3f} ns")
    print(f"Descida : {down_t0*1e9:.3f} ns -> {down_t1*1e9:.3f} ns")

    if down_t1 > t[-1]:
        print()
        print("AVISO:")
        print("A janela de descida ultrapassa o tempo final do ficheiro.")

    # ========================================================
    # EXTRAIR THRESHOLDS
    # ========================================================

    thresholds_up_all = []
    thresholds_down_all = []

    for comp in range(n_outputs):

        y = outs[:, comp]

        vth_up = interp_crossing_vin(
            t=t,
            vin=vin,
            y=y,
            threshold=VOUT_THRESHOLD,
            t0=up_t0,
            t1=up_t1,
            crossing_type="rising"
        )

        vth_down = interp_crossing_vin(
            t=t,
            vin=vin,
            y=y,
            threshold=VOUT_THRESHOLD,
            t0=down_t0,
            t1=down_t1,
            crossing_type="falling"
        )

        thresholds_up_all.append(vth_up)
        thresholds_down_all.append(vth_down)

    thresholds_up_all = np.array(thresholds_up_all)
    thresholds_down_all = np.array(thresholds_down_all)

    # ========================================================
    # HYSTERESIS
    # ========================================================

    hysteresis_v = thresholds_up_all - thresholds_down_all
    hysteresis_lsb = hysteresis_v / LSB

    thresholds_avg = 0.5 * (thresholds_up_all + thresholds_down_all)

    finite_avg = np.isfinite(thresholds_avg)
    thresholds_avg_sorted = np.sort(thresholds_avg[finite_avg])

    missing_thresholds = N_COMPARATORS - len(thresholds_avg_sorted)

    # ========================================================
    # DNL / INL / OFFSET
    # ========================================================

    dnl, inl, ideal_thresholds, widths = compute_dnl_inl(thresholds_avg_sorted)

    if dnl is None:
        print()
        print("ERRO: não foi possível calcular DNL/INL. Faltam thresholds.")
        passed = False

        max_abs_dnl_all = np.nan
        max_abs_dnl_internal = np.nan
        max_abs_dnl_used = np.nan

        worst_dnl_code = np.nan
        worst_dnl_value = np.nan

        max_abs_inl = np.nan
        worst_inl_transition = np.nan
        worst_inl_value = np.nan

        offset_v = np.nan
        offset_lsb = np.nan

    else:
        offset_v = np.mean(thresholds_avg_sorted - ideal_thresholds)
        offset_lsb = offset_v / LSB

        # DNL total incluindo code 0 e code 15
        max_abs_dnl_all = np.max(np.abs(dnl))

        # DNL interno ignorando endpoints
        dnl_internal = dnl[1:-1]
        max_abs_dnl_internal = np.max(np.abs(dnl_internal))

        # DNL usado para PASS/FAIL
        if USE_INTERNAL_DNL_FOR_PASS:
            max_abs_dnl_used = max_abs_dnl_internal
        else:
            max_abs_dnl_used = max_abs_dnl_all

        # Debug pior DNL
        worst_dnl_code = int(np.argmax(np.abs(dnl)))
        worst_dnl_value = dnl[worst_dnl_code]

        # INL
        max_abs_inl = np.max(np.abs(inl))
        worst_inl_transition = int(np.argmax(np.abs(inl))) + 1
        worst_inl_value = inl[worst_inl_transition - 1]

        passed = True

    if np.all(~np.isfinite(hysteresis_lsb)):
        max_abs_hyst_lsb = np.nan
        max_abs_hyst_v = np.nan
    else:
        max_abs_hyst_lsb = np.nanmax(np.abs(hysteresis_lsb))
        max_abs_hyst_v = max_abs_hyst_lsb * LSB

    # ========================================================
    # PASS / FAIL
    # ========================================================

    if missing_thresholds != 0:
        passed = False

    if np.isfinite(max_abs_dnl_used) and max_abs_dnl_used > MAX_ABS_DNL_LIMIT:
        passed = False

    if np.isfinite(max_abs_inl) and max_abs_inl > MAX_ABS_INL_LIMIT:
        passed = False

    if np.isfinite(max_abs_hyst_lsb) and max_abs_hyst_lsb > MAX_HYST_LIMIT:
        passed = False

    if np.isfinite(offset_lsb) and abs(offset_lsb) > MAX_OFFSET_LIMIT:
        passed = False

    # ========================================================
    # PRINT RESULTADOS
    # ========================================================

    print()
    print("------------- RESULTADOS ESTÁTICOS -------------")
    print(f"Thresholds encontrados          = {len(thresholds_avg_sorted)}/{N_COMPARATORS}")
    print(f"Missing thresholds              = {missing_thresholds}")
    print(f"Offset médio                    = {offset_v*1e3:.3f} mV")
    print(f"Offset médio                    = {offset_lsb:.3f} LSB")

    print()
    print("------------- DEBUG DNL / INL -------------")
    print(f"Pior código DNL                 = {worst_dnl_code}")
    print(f"DNL pior código                 = {worst_dnl_value:.3f} LSB")
    print(f"Max |DNL| incluindo extremos    = {max_abs_dnl_all:.3f} LSB")
    print(f"Max |DNL| interno code 1..14    = {max_abs_dnl_internal:.3f} LSB")
    print(f"Max |DNL| usado PASS/FAIL       = {max_abs_dnl_used:.3f} LSB")
    print(f"Pior transição INL              = {worst_inl_transition}")
    print(f"INL pior transição              = {worst_inl_value:.3f} LSB")

    print()
    print("------------- LIMITES -------------")
    print(f"Max |INL|                       = {max_abs_inl:.3f} LSB")
    print(f"Max |Hysteresis|                = {max_abs_hyst_v*1e3:.3f} mV")
    print(f"Max |Hysteresis|                = {max_abs_hyst_lsb:.3f} LSB")
    print(f"PASS/FAIL                       = {'PASS' if passed else 'FAIL'}")

    # ========================================================
    # TABELAS
    # ========================================================

    threshold_table = pd.DataFrame({
        "out_index": out_indices,
        "Vth_up_V": thresholds_up_all,
        "Vth_down_V": thresholds_down_all,
        "Vth_avg_V": thresholds_avg,
        "hysteresis_V": hysteresis_v,
        "hysteresis_LSB": hysteresis_lsb
    })

    threshold_table = threshold_table.sort_values("Vth_avg_V").reset_index(drop=True)

    if dnl is not None:
        dnl_table = pd.DataFrame({
            "code": np.arange(N_CODES),
            "width_V": widths,
            "DNL_LSB": dnl,
            "is_edge_code": [True] + [False] * (N_CODES - 2) + [True]
        })

        inl_table = pd.DataFrame({
            "transition": np.arange(1, N_COMPARATORS + 1),
            "Vth_avg_sorted_V": thresholds_avg_sorted,
            "Vth_ideal_V": ideal_thresholds,
            "INL_LSB": inl
        })
    else:
        dnl_table = pd.DataFrame()
        inl_table = pd.DataFrame()

    # ========================================================
    # PLOTS
    # ========================================================

    if make_plots:

        plt.figure(figsize=(10, 4))
        plt.plot(t * 1e9, vin, label="Vin")
        plt.axvspan(up_t0*1e9, up_t1*1e9, alpha=0.2, label="Subida")
        plt.axvspan(hold_t0*1e9, hold_t1*1e9, alpha=0.2, label="Hold")
        plt.axvspan(down_t0*1e9, down_t1*1e9, alpha=0.2, label="Descida")
        plt.title(f"Vin e janelas usadas - {filename}")
        plt.xlabel("Tempo [ns]")
        plt.ylabel("Vin [V]")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(10, 4))
        plt.plot(threshold_table["Vth_avg_V"], "o-", label="Threshold médio")
        if ideal_thresholds is not None:
            plt.plot(np.sort(ideal_thresholds), "x--", label="Threshold ideal")
        plt.title(f"Thresholds médios - {filename}")
        plt.xlabel("Comparador ordenado por threshold")
        plt.ylabel("Threshold [V]")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(10, 4))
        plt.stem(threshold_table["hysteresis_LSB"])
        plt.axhline(MAX_HYST_LIMIT, color="red", linestyle="--", label="+limite")
        plt.axhline(-MAX_HYST_LIMIT, color="red", linestyle="--", label="-limite")
        plt.title(f"Hysteresis por comparador - {filename}")
        plt.xlabel("Comparador ordenado por threshold")
        plt.ylabel("Hysteresis [LSB]")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        if dnl is not None:
            colors = ["red" if edge else "blue" for edge in dnl_table["is_edge_code"]]

            plt.figure(figsize=(10, 4))
            plt.bar(dnl_table["code"], dnl_table["DNL_LSB"], color=colors)
            plt.axhline(0, color="black", linewidth=1)
            plt.axhline(MAX_ABS_DNL_LIMIT, color="red", linestyle="--", linewidth=1)
            plt.axhline(-MAX_ABS_DNL_LIMIT, color="red", linestyle="--", linewidth=1)
            plt.title(f"DNL - {filename} | vermelho = endpoint code")
            plt.xlabel("Código")
            plt.ylabel("DNL [LSB]")
            plt.grid(True)
            plt.tight_layout()
            plt.show()

            plt.figure(figsize=(10, 4))
            plt.stem(inl_table["transition"], inl_table["INL_LSB"])
            plt.axhline(0, color="black", linewidth=1)
            plt.axhline(MAX_ABS_INL_LIMIT, color="red", linestyle="--", linewidth=1)
            plt.axhline(-MAX_ABS_INL_LIMIT, color="red", linestyle="--", linewidth=1)
            plt.title(f"INL - {filename}")
            plt.xlabel("Transição")
            plt.ylabel("INL [LSB]")
            plt.grid(True)
            plt.tight_layout()
            plt.show()

    return {
        "filename": filename,
        "ramp_ns": RAMP_TIME_NS,
        "fs_active": FS_ACTIVE,
        "fs_system": FS_SYSTEM,
        "passed": passed,
        "missing_thresholds": missing_thresholds,
        "offset_v": offset_v,
        "offset_lsb": offset_lsb,
        "max_abs_dnl_all": max_abs_dnl_all,
        "max_abs_dnl_internal": max_abs_dnl_internal,
        "max_abs_dnl_used": max_abs_dnl_used,
        "worst_dnl_code": worst_dnl_code,
        "worst_dnl_value": worst_dnl_value,
        "max_abs_inl": max_abs_inl,
        "worst_inl_transition": worst_inl_transition,
        "worst_inl_value": worst_inl_value,
        "max_abs_hyst_v": max_abs_hyst_v,
        "max_abs_hyst_lsb": max_abs_hyst_lsb,
        "threshold_table": threshold_table,
        "dnl_table": dnl_table,
        "inl_table": inl_table
    }


# ============================================================
# UPLOAD DOS FICHEIROS
# ============================================================

print()
print("===================================")
print("UPLOAD DOS FICHEIROS")
print("===================================")
print("Seleciona ficheiros tipo:")
print("  adc_ramp_35ns.txt")
print("  adc_ramp_30ns.txt")
print("  adc_ramp_25ns.txt")
print("  adc_ramp_10ns.txt")
print("  adc_ramp_5ns.txt")

uploaded = files.upload()
filenames = list(uploaded.keys())

# ============================================================
# ANALISAR TODOS OS FICHEIROS
# ============================================================

all_results = []

for filename in filenames:
    try:
        res = analyze_file(filename, make_plots=True)
        all_results.append(res)
    except Exception as e:
        print()
        print("===================================")
        print(f"ERRO AO ANALISAR: {filename}")
        print("===================================")
        print(str(e))


# ============================================================
# RESUMO GLOBAL
# ============================================================

summary_rows = []

for r in all_results:
    summary_rows.append({
        "filename": r["filename"],
        "ramp_ns": r["ramp_ns"],
        "fs_active_MHz": r["fs_active"] / 1e6,
        "fs_system_MHz": r["fs_system"] / 1e6,
        "passed": r["passed"],
        "offset_mV": r["offset_v"] * 1e3,
        "offset_LSB": r["offset_lsb"],
        "max_abs_DNL_all_LSB": r["max_abs_dnl_all"],
        "max_abs_DNL_internal_LSB": r["max_abs_dnl_internal"],
        "max_abs_DNL_used_LSB": r["max_abs_dnl_used"],
        "worst_DNL_code": r["worst_dnl_code"],
        "worst_DNL_value_LSB": r["worst_dnl_value"],
        "max_abs_INL_LSB": r["max_abs_inl"],
        "worst_INL_transition": r["worst_inl_transition"],
        "worst_INL_value_LSB": r["worst_inl_value"],
        "max_abs_hyst_mV": r["max_abs_hyst_v"] * 1e3,
        "max_abs_hyst_LSB": r["max_abs_hyst_lsb"],
        "missing_thresholds": r["missing_thresholds"]
    })

summary_df = pd.DataFrame(summary_rows)

if len(summary_df) > 0:
    summary_df = summary_df.sort_values("fs_active_MHz").reset_index(drop=True)

print()
print("===================================")
print("RESUMO GLOBAL")
print("===================================")

if len(summary_df) == 0:
    print("Nenhum resultado válido.")
else:
    print(summary_df.to_string(index=False))


# ============================================================
# MAIOR FS VÁLIDO
# ============================================================

if len(summary_df) > 0:

    passing = summary_df[summary_df["passed"] == True]

    print()
    print("===================================")
    print("MAIOR fs VÁLIDO")
    print("===================================")

    if len(passing) > 0:
        best = passing.sort_values("fs_active_MHz", ascending=False).iloc[0]

        print(f"Ficheiro                    = {best['filename']}")
        print(f"Ramp                        = {best['ramp_ns']:.3f} ns")
        print(f"fs ativo máximo             = {best['fs_active_MHz']:.3f} MHz")
        print(f"fs sistema com hold         = {best['fs_system_MHz']:.3f} MHz")
        print(f"Max |Hysteresis|            = {best['max_abs_hyst_LSB']:.3f} LSB")
        print(f"Max |DNL usado|             = {best['max_abs_DNL_used_LSB']:.3f} LSB")
        print(f"Max |DNL interno|           = {best['max_abs_DNL_internal_LSB']:.3f} LSB")
        print(f"Max |DNL all|               = {best['max_abs_DNL_all_LSB']:.3f} LSB")
        print(f"Max |INL|                   = {best['max_abs_INL_LSB']:.3f} LSB")
        print(f"Offset                      = {best['offset_LSB']:.3f} LSB")
    else:
        print("Nenhum ficheiro passou nos critérios definidos.")


# ============================================================
# PLOTS GLOBAIS
# ============================================================

if len(summary_df) > 0:

    plt.figure(figsize=(10, 5))
    plt.plot(summary_df["fs_active_MHz"], summary_df["max_abs_hyst_LSB"], "o-")
    plt.axhline(MAX_HYST_LIMIT, color="red", linestyle="--", label="Limite")
    plt.title("Hysteresis vs fs ativo")
    plt.xlabel("fs ativo [MHz]")
    plt.ylabel("Max |Hysteresis| [LSB]")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(summary_df["fs_active_MHz"], summary_df["max_abs_DNL_all_LSB"], "o-", label="Max |DNL| all")
    plt.plot(summary_df["fs_active_MHz"], summary_df["max_abs_DNL_internal_LSB"], "s-", label="Max |DNL| interno")
    plt.plot(summary_df["fs_active_MHz"], summary_df["max_abs_INL_LSB"], "^-", label="Max |INL|")
    plt.axhline(MAX_ABS_DNL_LIMIT, color="red", linestyle="--", label="Limite DNL")
    plt.axhline(MAX_ABS_INL_LIMIT, color="orange", linestyle="--", label="Limite INL")
    plt.title("DNL / INL vs fs ativo")
    plt.xlabel("fs ativo [MHz]")
    plt.ylabel("Erro [LSB]")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 5))
    colors = ["green" if p else "red" for p in summary_df["passed"]]
    plt.scatter(summary_df["fs_active_MHz"], summary_df["ramp_ns"], c=colors, s=100)
    plt.title("PASS/FAIL vs fs ativo")
    plt.xlabel("fs ativo [MHz]")
    plt.ylabel("Ramp time [ns]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import re

# ============================================================
# CONFIGURAÇÃO DO TESTE
# ============================================================

FIN = 9e6              # frequência da senoide [Hz]
FS_ANALYSIS = 180e6    # grelha uniforme para FFT [Hz]
                       # isto é apenas para análise, não é clock real do ADC

VOUT_THRESHOLD = 0.4   # threshold dos outputs digitais

V_MIN = 0.1
V_MAX = 0.7

N_BITS = 4
N_CODES = 2**N_BITS
N_COMPARATORS = N_CODES - 1

LSB = (V_MAX - V_MIN) / N_CODES

# Se quiseres ignorar transiente inicial, mete por exemplo 1e-6
T_START_IGNORE = 0.0

# Número máximo de harmónicos usados para THD/SNR
N_HARMONICS = 7

# Janela temporal mostrada nos plots
PLOT_WINDOW_US = 1.0

print("===================================")
print("CONFIGURAÇÃO")
print("===================================")
print(f"FIN              = {FIN/1e6:.3f} MHz")
print(f"FS_ANALYSIS      = {FS_ANALYSIS/1e6:.3f} MHz")
print(f"V_MIN            = {V_MIN:.3f} V")
print(f"V_MAX            = {V_MAX:.3f} V")
print(f"LSB              = {LSB*1e3:.3f} mV")
print(f"VOUT_THRESHOLD   = {VOUT_THRESHOLD:.3f} V")


# ============================================================
# FUNÇÕES
# ============================================================

def read_waveview_file(filename):
    """
    Lê ficheiro WaveView.

    Usa apenas:
        v(/out<0>) ... v(/out<14>)

    Ignora:
        v(out<0>) ... v(out<14>)
    """

    header = None
    header_line_idx = None

    with open(filename, "r") as f:
        for i, line in enumerate(f):
            s = line.strip()

            if s.startswith("XVAL"):
                header = s.split()
                header_line_idx = i
                break

    if header is None:
        raise ValueError("Header XVAL não encontrado.")

    out_dict = {}
    ignored_analog = []
    duplicate_out_info = []
    vin_col = None

    for i, name in enumerate(header):

        clean = (
            str(name)
            .replace("&lt;", "<")
            .replace("&gt;", ">")
            .strip()
            .lower()
        )

        # Vin
        if "v(/vin+" in clean:
            vin_col = i

        # Outputs digitais v(/out<k>)
        m_digital = re.search(r"v\(/out<(\d+)>\)", clean)

        if m_digital:
            idx = int(m_digital.group(1))

            if 0 <= idx <= 14:
                if idx not in out_dict:
                    out_dict[idx] = (i, idx, name)
                else:
                    duplicate_out_info.append((i, idx, name))

            continue

        # Outputs analógicos v(out<k>) ignorados
        m_analog = re.search(r"v\(out<(\d+)>\)", clean)

        if m_analog:
            ignored_analog.append((i, int(m_analog.group(1)), name))

    if vin_col is None:
        raise ValueError("Coluna Vin não encontrada.")

    if len(out_dict) == 0:
        raise ValueError("Nenhuma coluna digital v(/out<0>)...v(/out<14>) encontrada.")

    # Ordenar por out<0>, out<1>, ..., out<14>
    out_info_sorted = [out_dict[k] for k in sorted(out_dict.keys())]

    out_cols = [x[0] for x in out_info_sorted]
    out_indices = [x[1] for x in out_info_sorted]

    print()
    print("===================================")
    print("COLUNAS USADAS")
    print("===================================")
    print(f"Vin column                   = {vin_col}")
    print(f"Outputs digitais usados      = {out_indices}")
    print(f"Colunas digitais usadas      = {out_cols}")
    print(f"N outputs digitais           = {len(out_indices)}")

    if len(ignored_analog) > 0:
        ignored_indices = [x[1] for x in ignored_analog]
        print(f"Outputs analógicos ignorados = {ignored_indices}")

    if len(duplicate_out_info) > 0:
        duplicate_indices = [x[1] for x in duplicate_out_info]
        print(f"Outputs digitais duplicados ignorados = {duplicate_indices}")

    if len(out_indices) != N_COMPARATORS:
        print()
        print("AVISO:")
        print(f"Foram encontrados {len(out_indices)} outputs digitais.")
        print(f"Esperado: {N_COMPARATORS}")

    # Ler todas as colunas para evitar problemas de ordem
    df = pd.read_csv(
        filename,
        sep=r"\s+",
        comment="#",
        skiprows=header_line_idx + 1,
        header=None,
        engine="c"
    )

    data = df.to_numpy(dtype=float)

    t = data[:, 0]
    vin = data[:, vin_col]
    outs = data[:, out_cols]

    # ordenar por tempo
    idx_sort = np.argsort(t)

    t = t[idx_sort]
    vin = vin[idx_sort]
    outs = outs[idx_sort, :]

    # remover tempos duplicados
    t_unique, unique_idx = np.unique(t, return_index=True)

    t = t_unique
    vin = vin[unique_idx]
    outs = outs[unique_idx, :]

    return t, vin, outs, out_indices


def alias_frequency(f, fs):
    """
    Calcula frequência aliased para dentro de [0, fs/2].
    Útil para harmónicos acima de Nyquist da grelha de análise.
    """

    f_mod = np.mod(f, fs)

    if f_mod > fs / 2:
        f_alias = fs - f_mod
    else:
        f_alias = f_mod

    return f_alias


def calculate_fft_metrics(x, fs, fin, n_harmonics=7):
    """
    Calcula SINAD, ENOB, SNR, THD e SFDR.
    """

    x = np.asarray(x)

    # remover DC
    x = x - np.mean(x)

    N = len(x)

    # verificar coerência
    fund_bin_float = fin / fs * N
    fund_bin = int(round(fund_bin_float))

    coherent = abs(fund_bin_float - fund_bin) < 0.05

    print()
    print("===================================")
    print("COERÊNCIA FFT")
    print("===================================")
    print(f"N amostras FFT          = {N}")
    print(f"Bin fundamental ideal   = {fund_bin_float:.3f}")
    print(f"Bin fundamental usado   = {fund_bin}")
    print(f"Coerente                = {coherent}")

    if coherent:
        window = np.ones(N)
    else:
        print("AVISO: não está perfeitamente coerente. A usar janela Hann.")
        window = np.hanning(N)

    xw = x * window

    X = np.fft.rfft(xw)
    P = np.abs(X)**2
    P = P / np.sum(window**2)

    # remover DC
    P[0] = 0

    freq = np.fft.rfftfreq(N, d=1/fs)

    if fund_bin <= 0 or fund_bin >= len(P):
        fund_bin = np.argmax(P[1:]) + 1

    # bins da fundamental
    if coherent:
        signal_bins = [fund_bin]
    else:
        signal_bins = [fund_bin - 1, fund_bin, fund_bin + 1]
        signal_bins = [b for b in signal_bins if 0 <= b < len(P)]

    signal_power = np.sum(P[signal_bins])

    # ========================================================
    # Harmónicos
    # ========================================================

    harmonic_bins = []

    for h in range(2, n_harmonics + 1):

        f_h = alias_frequency(h * fin, fs)
        hb = int(round(f_h / (fs / N)))

        if hb <= 0 or hb >= len(P):
            continue

        # evitar contar a fundamental como harmónico por alias
        if hb in signal_bins:
            continue

        if hb not in harmonic_bins:
            harmonic_bins.append(hb)

    harmonic_power = np.sum(P[harmonic_bins]) if len(harmonic_bins) > 0 else 0.0

    # ========================================================
    # SINAD
    # fundamental / tudo o resto exceto DC
    # ========================================================

    noise_dist_mask = np.ones(len(P), dtype=bool)
    noise_dist_mask[0] = False

    for b in signal_bins:
        noise_dist_mask[b] = False

    noise_dist_power = np.sum(P[noise_dist_mask])

    SINAD = 10 * np.log10(signal_power / noise_dist_power)
    ENOB = (SINAD - 1.76) / 6.02

    # ========================================================
    # SNR
    # fundamental / ruído, excluindo harmónicos
    # ========================================================

    noise_only_mask = np.ones(len(P), dtype=bool)
    noise_only_mask[0] = False

    for b in signal_bins:
        noise_only_mask[b] = False

    for b in harmonic_bins:
        noise_only_mask[b] = False

    noise_only_power = np.sum(P[noise_only_mask])

    SNR = 10 * np.log10(signal_power / noise_only_power) if noise_only_power > 0 else np.inf

    # ========================================================
    # THD
    # harmónicos / fundamental
    # ========================================================

    THD = 10 * np.log10(harmonic_power / signal_power) if harmonic_power > 0 else -np.inf

    # ========================================================
    # SFDR
    # fundamental / maior spur
    # exclui DC e fundamental
    # ========================================================

    spur_mask = np.ones(len(P), dtype=bool)
    spur_mask[0] = False

    for b in signal_bins:
        spur_mask[b] = False

    spur_power = np.max(P[spur_mask])
    spur_bin = np.argmax(np.where(spur_mask, P, -np.inf))

    SFDR = 10 * np.log10(signal_power / spur_power)

    return {
        "N": N,
        "freq": freq,
        "P": P,
        "fund_bin": fund_bin,
        "fund_freq": freq[fund_bin],
        "signal_bins": signal_bins,
        "harmonic_bins": harmonic_bins,
        "SINAD": SINAD,
        "ENOB": ENOB,
        "SNR": SNR,
        "THD": THD,
        "SFDR": SFDR,
        "spur_bin": spur_bin,
        "spur_freq": freq[spur_bin],
        "coherent": coherent
    }


def calculate_bubble_rate(bits):
    """
    Calcula bubble error rate para código termómetro.

    bits: matriz [N_amostras x N_comparadores]
          Cada linha contém os bits termómetro ordenados por out<0>...out<14>.

    Um padrão é considerado válido se for monotónico:
      - 111110000000000
      - 000001111111111

    Bubble error ocorre se houver mais de uma transição entre 0 e 1.
    """

    bits = np.asarray(bits).astype(int)

    n_samples = bits.shape[0]

    bubble_flags = []
    n_transitions_list = []
    bubble_strength_list = []

    for row in bits:

        diff = np.diff(row)

        # contar transições 0->1 ou 1->0
        transitions = np.sum(diff != 0)

        # padrão termómetro ideal tem no máximo 1 transição
        has_bubble = transitions > 1

        bubble_flags.append(has_bubble)
        n_transitions_list.append(transitions)

        # força do bubble:
        # 0 se não houver bubble
        # transitions - 1 se houver múltiplas quebras
        bubble_strength_list.append(max(0, transitions - 1))

    bubble_flags = np.array(bubble_flags)
    n_transitions_list = np.array(n_transitions_list)
    bubble_strength_list = np.array(bubble_strength_list)

    n_bubble_samples = np.sum(bubble_flags)

    bubble_rate = n_bubble_samples / n_samples
    bubble_rate_percent = 100 * bubble_rate

    mean_bubble_strength = np.mean(bubble_strength_list)
    max_bubble_strength = np.max(bubble_strength_list)

    return {
        "n_samples": n_samples,
        "n_bubble_samples": n_bubble_samples,
        "bubble_rate": bubble_rate,
        "bubble_rate_percent": bubble_rate_percent,
        "mean_bubble_strength": mean_bubble_strength,
        "max_bubble_strength": max_bubble_strength,
        "bubble_flags": bubble_flags,
        "n_transitions": n_transitions_list,
        "bubble_strength": bubble_strength_list
    }


# ============================================================
# UPLOAD DO FICHEIRO
# ============================================================

print()
print("===================================")
print("UPLOAD DO FICHEIRO")
print("===================================")

uploaded = files.upload()
filename = list(uploaded.keys())[0]

print(f"Ficheiro carregado: {filename}")

# ============================================================
# LER DADOS
# ============================================================

t_raw, vin_raw, outs_raw, out_indices = read_waveview_file(filename)

print()
print("===================================")
print("DADOS LIDOS")
print("===================================")
print(f"N pontos brutos    = {len(t_raw)}")
print(f"Tempo inicial      = {t_raw[0]*1e6:.3f} us")
print(f"Tempo final        = {t_raw[-1]*1e6:.3f} us")

# ============================================================
# CORTAR TRANSIENTE INICIAL
# ============================================================

mask_time = t_raw >= (t_raw[0] + T_START_IGNORE)

t_raw = t_raw[mask_time]
vin_raw = vin_raw[mask_time]
outs_raw = outs_raw[mask_time, :]

# ============================================================
# CRIAR GRELHA UNIFORME PARA ANÁLISE FFT
# ============================================================

t_start = t_raw[0]
t_stop = t_raw[-1]

t_sample = np.arange(t_start, t_stop, 1 / FS_ANALYSIS)

if len(t_sample) < 100:
    raise ValueError("Poucas amostras na grelha uniforme. Aumenta tempo de simulação ou reduz FS_ANALYSIS.")

print()
print("===================================")
print("GRELHA DE ANÁLISE")
print("===================================")
print(f"FS_ANALYSIS        = {FS_ANALYSIS/1e6:.3f} MHz")
print(f"N amostras         = {len(t_sample)}")
print(f"Tempo útil         = {(t_sample[-1]-t_sample[0])*1e6:.3f} us")

# ============================================================
# INTERPOLAR VIN E OUTPUTS DIGITAIS
# ============================================================

vin_sampled = np.interp(t_sample, t_raw, vin_raw)

outs_sampled = np.zeros((len(t_sample), outs_raw.shape[1]))

for k in range(outs_raw.shape[1]):
    outs_sampled[:, k] = np.interp(t_sample, t_raw, outs_raw[:, k])

# ============================================================
# CONVERTER OUTPUTS PARA BITS
# ============================================================

bits = (outs_sampled > VOUT_THRESHOLD).astype(int)

# Código termómetro = número de 1s
adc_code = np.sum(bits, axis=1)

adc_code = np.clip(adc_code, 0, N_COMPARATORS)

# ============================================================
# BUBBLE ERROR RATE
# ============================================================

bubble = calculate_bubble_rate(bits)

# ============================================================
# RECONSTRUIR VALOR ANALÓGICO EQUIVALENTE
# ============================================================

adc_reconstructed = V_MIN + (adc_code / N_COMPARATORS) * (V_MAX - V_MIN)

# ============================================================
# CÁLCULO FFT / SINAD / ENOB / SNR / THD / SFDR
# ============================================================

metrics = calculate_fft_metrics(
    x=adc_reconstructed,
    fs=FS_ANALYSIS,
    fin=FIN,
    n_harmonics=N_HARMONICS
)

# ============================================================
# RESULTADOS
# ============================================================

print()
print("===================================")
print("RESULTADOS DINÂMICOS")
print("===================================")
print(f"fin                  = {FIN/1e6:.3f} MHz")
print(f"FS análise           = {FS_ANALYSIS/1e6:.3f} MHz")
print(f"N FFT                = {metrics['N']}")
print(f"Fundamental bin      = {metrics['fund_bin']}")
print(f"Fundamental freq     = {metrics['fund_freq']/1e6:.6f} MHz")
print(f"SINAD                = {metrics['SINAD']:.2f} dB")
print(f"ENOB                 = {metrics['ENOB']:.2f} bits")
print(f"SNR                  = {metrics['SNR']:.2f} dB")
print(f"THD                  = {metrics['THD']:.2f} dB")
print(f"SFDR                 = {metrics['SFDR']:.2f} dB")
print(f"Maior spur           = {metrics['spur_freq']/1e6:.6f} MHz")
print(f"Código mínimo        = {np.min(adc_code)}")
print(f"Código máximo        = {np.max(adc_code)}")
print(f"Códigos usados       = {np.unique(adc_code)}")

print()
print("===================================")
print("BUBBLE ERRORS")
print("===================================")
print(f"Amostras analisadas       = {bubble['n_samples']}")
print(f"Amostras com bubble       = {bubble['n_bubble_samples']}")
print(f"Bubble rate               = {bubble['bubble_rate_percent']:.4f} %")
print(f"Bubble strength médio     = {bubble['mean_bubble_strength']:.4f}")
print(f"Bubble strength máximo    = {bubble['max_bubble_strength']}")

# Slew rate teórico da senoide
VIN_AMPLITUDE_EST = (np.max(vin_sampled) - np.min(vin_sampled)) / 2
VIN_OFFSET_EST = (np.max(vin_sampled) + np.min(vin_sampled)) / 2

SR_THEORETICAL = 2 * np.pi * FIN * VIN_AMPLITUDE_EST
SR_THEORETICAL_V_PER_US = SR_THEORETICAL / 1e6

print()
print("===================================")
print("SENOIDE / SLEW RATE")
print("===================================")
print(f"Amplitude estimada   = {VIN_AMPLITUDE_EST:.3f} V")
print(f"Offset estimado      = {VIN_OFFSET_EST:.3f} V")
print(f"SR máximo estimado   = {SR_THEORETICAL_V_PER_US:.3f} V/us")

# ============================================================
# PLOTS TEMPORAIS — mostrar apenas 1 us
# ============================================================

t_us = t_sample * 1e6
t0_us = t_us[0]
t1_us = t0_us + PLOT_WINDOW_US

plot_mask = (t_us >= t0_us) & (t_us <= t1_us)

plt.figure(figsize=(12,4))
plt.plot(t_us[plot_mask], vin_sampled[plot_mask], label="Vin")
plt.plot(t_us[plot_mask], adc_reconstructed[plot_mask], label="ADC reconstruído", alpha=0.8)
plt.title("Vin vs ADC reconstruído — primeiros 1 µs")
plt.xlabel("Tempo [µs]")
plt.ylabel("Tensão [V]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plt.plot(
    t_us[plot_mask],
    adc_code[plot_mask],
    marker='.',
    linestyle='-',
    markersize=3
)
plt.title("Código ADC reconstruído — primeiros 1 µs")
plt.xlabel("Tempo [µs]")
plt.ylabel("Código")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plt.plot(
    t_us[plot_mask],
    bubble["bubble_flags"][plot_mask].astype(int),
    marker='.',
    linestyle='-',
    markersize=3
)
plt.title("Bubble errors — primeiros 1 µs")
plt.xlabel("Tempo [µs]")
plt.ylabel("Bubble error")
plt.yticks([0, 1], ["No bubble", "Bubble"])
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================================
# PLOT FFT
# ============================================================

plt.figure(figsize=(12,5))
plt.plot(metrics["freq"] / 1e6, 10*np.log10(metrics["P"] + 1e-30))
plt.axvline(FIN/1e6, color="red", linestyle="--", label="Fundamental")
plt.axvline(metrics["spur_freq"]/1e6, color="orange", linestyle="--", label="Maior spur")
plt.title("FFT do ADC reconstruído")
plt.xlabel("Frequência [MHz]")
plt.ylabel("Potência [dB]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ============================================================
# HISTOGRAMA DOS CÓDIGOS
# ============================================================

plt.figure(figsize=(8,4))
plt.hist(adc_code, bins=np.arange(-0.5, N_COMPARATORS + 1.5, 1), edgecolor="black")
plt.title("Histograma dos códigos ADC")
plt.xlabel("Código")
plt.ylabel("Contagem")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import re

# ============================================================
# CONFIGURAÇÃO
# ============================================================

V_MIN = 0.1
V_MAX = 0.7
N_BITS = 4
N_CODES = 2**N_BITS

LSB = (V_MAX - V_MIN) / N_CODES

print("===================================")
print("CONFIGURAÇÃO")
print("===================================")
print(f"V_MIN = {V_MIN:.3f} V")
print(f"V_MAX = {V_MAX:.3f} V")
print(f"N_BITS = {N_BITS}")
print(f"N_CODES = {N_CODES}")
print(f"LSB = {LSB*1e3:.3f} mV")


# ============================================================
# FUNÇÃO PARA LER FICHEIRO WAVEVIEW DE THRESHOLD MC
# ============================================================

def read_mc_threshold_file(filename, max_mc=330):
    """
    Lê um ficheiro WaveView exportado com:
      yvalue(v(vin), cross(v(outX), 0.4, 1)):MONTE_CARLO=n

    Retorna:
      mc_numbers: array dos números Monte Carlo
      values: thresholds correspondentes
    """

    header = None
    header_line_idx = None

    with open(filename, "r") as f:
        for i, line in enumerate(f):
            s = line.strip()
            if s.startswith("XVAL"):
                header = s.split()
                header_line_idx = i
                break

    if header is None:
        raise ValueError(f"Header XVAL não encontrado em {filename}")

    # Identificar colunas Monte Carlo válidas
    mc_cols = []
    mc_nums = []

    for col_idx, name in enumerate(header):
        m = re.search(r"MONTE_CARLO=(\d+)", name)

        if m:
            mc = int(m.group(1))

            # Ignorar MONTE_CARLO=0 e qualquer extra acima de max_mc
            if 1 <= mc <= max_mc:
                mc_cols.append(col_idx)
                mc_nums.append(mc)

    if len(mc_cols) == 0:
        raise ValueError(f"Nenhuma coluna Monte Carlo válida encontrada em {filename}")

    # Ler dados
    df = pd.read_csv(
        filename,
        sep=r"\s+",
        comment="#",
        skiprows=header_line_idx + 1,
        header=None,
        na_values=["NoData", "NAN", "NaN", "nan"],
        engine="python"
    )

    data = df.to_numpy(dtype=float)

    # Normalmente há uma ou mais linhas; usamos a primeira linha com dados válidos
    # Coluna 0 é XVAL; colunas seguintes correspondem ao header.
    # Como header inclui XVAL na posição 0, usamos diretamente mc_cols.
    values_matrix = data[:, mc_cols]

    # Se houver várias linhas, escolher a primeira linha onde exista algum valor válido
    valid_row_found = False

    for r in range(values_matrix.shape[0]):
        row = values_matrix[r, :]
        if np.sum(np.isfinite(row)) > 0:
            values = row
            valid_row_found = True
            break

    if not valid_row_found:
        raise ValueError(f"Nenhuma linha com dados válidos encontrada em {filename}")

    mc_nums = np.array(mc_nums)
    values = np.array(values, dtype=float)

    # Ordenar por número de Monte Carlo
    idx_sort = np.argsort(mc_nums)

    mc_nums = mc_nums[idx_sort]
    values = values[idx_sort]

    return mc_nums, values


# ============================================================
# UPLOAD DOS DOIS FICHEIROS
# ============================================================

print()
print("===================================")
print("UPLOAD")
print("===================================")
print("Seleciona os dois ficheiros:")
print("  1) threshold de out0")
print("  2) threshold de out1")

uploaded = files.upload()
filenames = list(uploaded.keys())

if len(filenames) != 2:
    raise ValueError("Faz upload exatamente de 2 ficheiros: out0 e out1.")

print()
print("Ficheiros carregados:")
for f in filenames:
    print(f"  {f}")


# ============================================================
# ESCOLHER QUAL É OUT0 E QUAL É OUT1 PELO NOME
# ============================================================

file_out0 = None
file_out1 = None

for f in filenames:
    fl = f.lower()

    if "out0" in fl or "out_0" in fl:
        file_out0 = f

    if "out1" in fl or "out_1" in fl:
        file_out1 = f

if file_out0 is None or file_out1 is None:
    print()
    print("AVISO:")
    print("Não consegui identificar automaticamente out0/out1 pelo nome.")
    print("Vou assumir:")
    print(f"  out0 = {filenames[0]}")
    print(f"  out1 = {filenames[1]}")
    file_out0 = filenames[0]
    file_out1 = filenames[1]

print()
print("===================================")
print("FICHEIROS USADOS")
print("===================================")
print(f"out0 = {file_out0}")
print(f"out1 = {file_out1}")


# ============================================================
# LER THRESHOLDS
# ============================================================

mc0, vth_out0 = read_mc_threshold_file(file_out0, max_mc=330)
mc1, vth_out1 = read_mc_threshold_file(file_out1, max_mc=330)

# Garantir que os MCs coincidem
common_mc = np.intersect1d(mc0, mc1)

v0 = []
v1 = []

for mc in common_mc:
    idx0 = np.where(mc0 == mc)[0][0]
    idx1 = np.where(mc1 == mc)[0][0]

    v0.append(vth_out0[idx0])
    v1.append(vth_out1[idx1])

v0 = np.array(v0, dtype=float)
v1 = np.array(v1, dtype=float)

# ============================================================
# CALCULAR DNL ENTRE OUT1 E OUT0
# ============================================================

valid = np.isfinite(v0) & np.isfinite(v1)

# Como out1 vem antes de out0:
width = v0[valid] - v1[valid]

dnl = width / LSB - 1

valid_mc = common_mc[valid]

# Detectar casos problemáticos
negative_width = width <= 0
missing_code = dnl <= -1

# ============================================================
# RESULTADOS
# ============================================================

print()
print("===================================")
print("RESULTADOS DNL ENTRE out1 E out0")
print("===================================")

print(f"Runs MC totais comuns        = {len(common_mc)}")
print(f"Runs válidos                 = {np.sum(valid)}")
print(f"Runs inválidos/NoData        = {len(common_mc) - np.sum(valid)}")
print(f"Yield válido                 = {100*np.mean(valid):.2f} %")

print()
print("------------- THRESHOLDS -------------")
print(f"Vth(out0) médio              = {np.nanmean(v0)*1e3:.3f} mV")
print(f"Vth(out1) médio              = {np.nanmean(v1)*1e3:.3f} mV")
print(f"Width médio out1→out0        = {np.mean(width)*1e3:.3f} mV")
print(f"Width sigma                  = {np.std(width)*1e3:.3f} mV")

print()
print("------------- DNL -------------")
print(f"DNL médio                    = {np.mean(dnl):.3f} LSB")
print(f"DNL sigma                    = {np.std(dnl):.3f} LSB")
print(f"DNL mínimo                   = {np.min(dnl):.3f} LSB")
print(f"DNL máximo                   = {np.max(dnl):.3f} LSB")
print(f"Max |DNL|                    = {np.max(np.abs(dnl)):.3f} LSB")

print()
print("------------- YIELD -------------")
print(f"Runs com width <= 0          = {np.sum(negative_width)}")
print(f"Runs com DNL <= -1           = {np.sum(missing_code)}")
print(f"Yield sem missing code       = {100*np.mean(~missing_code):.2f} %")
print(f"Yield |DNL| <= 0.5 LSB       = {100*np.mean(np.abs(dnl) <= 0.5):.2f} %")
print(f"Yield |DNL| <= 1.0 LSB       = {100*np.mean(np.abs(dnl) <= 1.0):.2f} %")


# ============================================================
# TABELA POR RUN
# ============================================================

results_df = pd.DataFrame({
    "MC_run": valid_mc,
    "Vth_out1_V": v1[valid],
    "Vth_out0_V": v0[valid],
    "width_V": width,
    "DNL_LSB": dnl,
    "missing_code": missing_code,
    "negative_width": negative_width
})

print()
print("Primeiras linhas:")
print(results_df.head(10).to_string(index=False))


# ============================================================
# PLOTS
# ============================================================

# ============================================================
# PLOTS - GALERIA DE ESTILOS
# ============================================================

# Escolhe o modo:
# "all"        -> mostra todos os plots
# "dnl"        -> só plots DNL vs Monte Carlo
# "hist"       -> só histogramas de DNL
# "width"      -> só plots Width vs Monte Carlo
# "thresholds" -> só plots Vth(out0) vs Vth(out1)
PLOT_MODE = "all"


# ============================================================
# FUNÇÕES AUXILIARES
# ============================================================

def add_dnl_limits():
    """
    Adiciona linhas de referência para gráficos de DNL.
    """
    plt.axhline(0, color="black", linewidth=1, label="0 LSB")
    plt.axhline(0.5, color="red", linestyle="--", linewidth=1.2, label="+0.5 LSB")
    plt.axhline(-0.5, color="red", linestyle="--", linewidth=1.2, label="-0.5 LSB")
    plt.axhline(-1.0, color="orange", linestyle="--", linewidth=1.2, label="Missing code")


def add_dnl_hist_limits():
    """
    Adiciona linhas de referência para histogramas de DNL.
    """
    plt.axvline(0, color="black", linewidth=1, label="0 LSB")
    plt.axvline(0.5, color="red", linestyle="--", linewidth=1.2, label="+0.5 LSB")
    plt.axvline(-0.5, color="red", linestyle="--", linewidth=1.2, label="-0.5 LSB")
    plt.axvline(-1.0, color="orange", linestyle="--", linewidth=1.2, label="Missing code")


def finish_plot(title, xlabel, ylabel):
    """
    Aplica labels, título, grelha, legenda e layout.
    """
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.35)
    plt.legend()
    plt.tight_layout()
    plt.show()


# ============================================================
# 1) HISTOGRAMAS DE DNL
# Eixo X: DNL [LSB]
# Eixo Y: Número de runs
# ============================================================

if PLOT_MODE in ["all", "hist"]:

    # ------------------------------------------------------------
    # HIST_01 - Histograma clássico
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.hist(dnl, bins=30, edgecolor="black", color="steelblue")
    add_dnl_hist_limits()
    finish_plot(
        "HIST_01 - Monte Carlo DNL, histograma clássico",
        "DNL [LSB]",
        "Número de runs"
    )


    # ------------------------------------------------------------
    # HIST_02 - Histograma verde com transparência
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.hist(dnl, bins=30, edgecolor="black", color="mediumseagreen", alpha=0.75)
    add_dnl_hist_limits()
    finish_plot(
        "HIST_02 - Distribuição DNL com transparência",
        "DNL [LSB]",
        "Número de runs"
    )


    # ------------------------------------------------------------
    # HIST_03 - Histograma laranja mais detalhado
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.hist(dnl, bins=45, edgecolor="black", color="darkorange", alpha=0.85)
    add_dnl_hist_limits()
    finish_plot(
        "HIST_03 - DNL com mais bins para maior detalhe",
        "DNL [LSB]",
        "Número de runs"
    )


    # ------------------------------------------------------------
    # HIST_04 - Histograma normalizado em densidade
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.hist(dnl, bins=30, density=True, edgecolor="black", color="mediumpurple", alpha=0.75)
    add_dnl_hist_limits()
    finish_plot(
        "HIST_04 - Distribuição DNL normalizada",
        "DNL [LSB]",
        "Densidade"
    )


# ============================================================
# 2) DNL POR MONTE CARLO RUN
# Eixo X: Monte Carlo run
# Eixo Y: DNL [LSB]
# ============================================================

if PLOT_MODE in ["all", "dnl"]:

    # ------------------------------------------------------------
    # DNL_01 - Scatter simples
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.plot(valid_mc, dnl, marker=".", linestyle="none", color="royalblue", label="DNL")
    add_dnl_limits()
    finish_plot(
        "DNL_01 - DNL por Monte Carlo run, pontos simples",
        "Monte Carlo run",
        "DNL [LSB]"
    )


    # ------------------------------------------------------------
    # DNL_02 - Scatter com pontos maiores e transparência
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.scatter(valid_mc, dnl, s=35, color="crimson", alpha=0.65, label="DNL")
    add_dnl_limits()
    finish_plot(
        "DNL_02 - DNL por run com pontos vermelhos transparentes",
        "Monte Carlo run",
        "DNL [LSB]"
    )


    # ------------------------------------------------------------
    # DNL_03 - Linha com marcadores
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.plot(valid_mc, dnl, marker="o", markersize=3, linestyle="-", linewidth=0.8,
             color="darkgreen", label="DNL")
    add_dnl_limits()
    finish_plot(
        "DNL_03 - DNL por run ligado por linha",
        "Monte Carlo run",
        "DNL [LSB]"
    )


    # ------------------------------------------------------------
    # DNL_04 - Stem plot
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    markerline, stemlines, baseline = plt.stem(valid_mc, dnl, label="DNL")
    plt.setp(markerline, marker="o", markersize=3, color="navy")
    plt.setp(stemlines, color="navy", linewidth=0.7, alpha=0.7)
    plt.setp(baseline, color="black", linewidth=0.8)
    add_dnl_limits()
    finish_plot(
        "DNL_04 - DNL por run em formato stem",
        "Monte Carlo run",
        "DNL [LSB]"
    )


    # ------------------------------------------------------------
    # DNL_05 - Barras verticais
    # ------------------------------------------------------------
    plt.figure(figsize=(12, 5))
    plt.bar(valid_mc, dnl, color="teal", alpha=0.75, label="DNL")
    add_dnl_limits()
    finish_plot(
        "DNL_05 - DNL por run em barras",
        "Monte Carlo run",
        "DNL [LSB]"
    )


    # ------------------------------------------------------------
    # DNL_06 - Cores dependentes do valor de DNL
    # ------------------------------------------------------------
    colors = np.where(np.abs(dnl) <= 0.5, "green",
             np.where(dnl <= -1.0, "orange", "red"))

    plt.figure(figsize=(10, 5))
    plt.scatter(valid_mc, dnl, s=35, c=colors, alpha=0.8, label="DNL")
    add_dnl_limits()
    finish_plot(
        "DNL_06 - DNL por run com cores por critério",
        "Monte Carlo run",
        "DNL [LSB]"
    )


    # ------------------------------------------------------------
    # DNL_07 - DNL com média e zona ± sigma
    # ------------------------------------------------------------
    dnl_mean = np.mean(dnl)
    dnl_std = np.std(dnl)

    plt.figure(figsize=(10, 5))
    plt.plot(valid_mc, dnl, marker=".", linestyle="none", color="royalblue", label="DNL")
    plt.axhline(dnl_mean, color="purple", linewidth=2, label=f"Média = {dnl_mean:.3f} LSB")
    plt.axhspan(dnl_mean - dnl_std, dnl_mean + dnl_std,
                color="purple", alpha=0.15, label=f"±1 sigma = {dnl_std:.3f} LSB")
    add_dnl_limits()
    finish_plot(
        "DNL_07 - DNL com média e banda de 1 sigma",
        "Monte Carlo run",
        "DNL [LSB]"
    )


# ============================================================
# 3) WIDTH POR MONTE CARLO RUN
# Eixo X: Monte Carlo run
# Eixo Y: Width [mV]
# ============================================================

if PLOT_MODE in ["all", "width"]:

    width_mV = width * 1e3
    LSB_mV = LSB * 1e3

    # ------------------------------------------------------------
    # WIDTH_01 - Scatter clássico
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.plot(valid_mc, width_mV, marker=".", linestyle="none", color="royalblue", label="Width")
    plt.axhline(LSB_mV, color="red", linestyle="--", linewidth=1.2, label="1 LSB ideal")
    finish_plot(
        "WIDTH_01 - Width entre out1 e out0 por run",
        "Monte Carlo run",
        "Width [mV]"
    )


    # ------------------------------------------------------------
    # WIDTH_02 - Linha com marcadores
    # ------------------------------------------------------------
    plt.figure(figsize=(10, 5))
    plt.plot(valid_mc, width_mV, marker="o", markersize=3, linestyle="-",
             linewidth=0.8, color="darkorange", label="Width")
    plt.axhline(LSB_mV, color="red", linestyle="--", linewidth=1.2, label="1 LSB ideal")
    finish_plot(
        "WIDTH_02 - Width por run ligado por linha",
        "Monte Carlo run",
        "Width [mV]"
    )


    # ------------------------------------------------------------
    # WIDTH_03 - Barras
    # ------------------------------------------------------------
    plt.figure(figsize=(12, 5))
    plt.bar(valid_mc, width_mV, color="mediumseagreen", alpha=0.75, label="Width")
    plt.axhline(LSB_mV, color="red", linestyle="--", linewidth=1.2, label="1 LSB ideal")
    finish_plot(
        "WIDTH_03 - Width por run em barras",
        "Monte Carlo run",
        "Width [mV]"
    )


    # ------------------------------------------------------------
    # WIDTH_04 - Width com média e sigma
    # ------------------------------------------------------------
    width_mean = np.mean(width_mV)
    width_std = np.std(width_mV)

    plt.figure(figsize=(10, 5))
    plt.scatter(valid_mc, width_mV, s=30, color="navy", alpha=0.7, label="Width")
    plt.axhline(LSB_mV, color="red", linestyle="--", linewidth=1.2, label="1 LSB ideal")
    plt.axhline(width_mean, color="purple", linewidth=2,
                label=f"Média = {width_mean:.3f} mV")
    plt.axhspan(width_mean - width_std, width_mean + width_std,
                color="purple", alpha=0.15,
                label=f"±1 sigma = {width_std:.3f} mV")
    finish_plot(
        "WIDTH_04 - Width com média e banda de 1 sigma",
        "Monte Carlo run",
        "Width [mV]"
    )


# ============================================================
# 4) Vth(out0) VS Vth(out1)
# Eixo X: Vth(out1) [mV]
# Eixo Y: Vth(out0) [mV]
# ============================================================

if PLOT_MODE in ["all", "thresholds"]:

    v1_mV = v1[valid] * 1e3
    v0_mV = v0[valid] * 1e3

    # ------------------------------------------------------------
    # TH_01 - Scatter simples
    # ------------------------------------------------------------
    plt.figure(figsize=(7, 6))
    plt.scatter(v1_mV, v0_mV, s=20, color="royalblue", alpha=0.75, label="Thresholds")
    finish_plot(
        "TH_01 - Vth(out0) vs Vth(out1), scatter clássico",
        "Vth(out1) [mV]",
        "Vth(out0) [mV]"
    )


    # ------------------------------------------------------------
    # TH_02 - Scatter com mapa de cor baseado no DNL
    # ------------------------------------------------------------
    plt.figure(figsize=(7, 6))
    sc = plt.scatter(v1_mV, v0_mV, s=30, c=dnl, cmap="coolwarm", alpha=0.85)
    plt.colorbar(sc, label="DNL [LSB]")
    plt.xlabel("Vth(out1) [mV]")
    plt.ylabel("Vth(out0) [mV]")
    plt.title("TH_02 - Vth(out0) vs Vth(out1) colorido por DNL")
    plt.grid(True, alpha=0.35)
    plt.tight_layout()
    plt.show()


    # ------------------------------------------------------------
    # TH_03 - Scatter com linha de referência y=x
    # ------------------------------------------------------------
    plt.figure(figsize=(7, 6))
    plt.scatter(v1_mV, v0_mV, s=25, color="darkgreen", alpha=0.75, label="Thresholds")

    min_v = min(np.min(v1_mV), np.min(v0_mV))
    max_v = max(np.max(v1_mV), np.max(v0_mV))
    plt.plot([min_v, max_v], [min_v, max_v], color="black", linestyle="--", label="y = x")

    finish_plot(
        "TH_03 - Vth(out0) vs Vth(out1) com linha y=x",
        "Vth(out1) [mV]",
        "Vth(out0) [mV]"
    )


    # ------------------------------------------------------------
    # TH_04 - Hexbin, útil quando há muitos pontos
    # ------------------------------------------------------------
    plt.figure(figsize=(7, 6))
    hb = plt.hexbin(v1_mV, v0_mV, gridsize=25, cmap="viridis", mincnt=1)
    plt.colorbar(hb, label="Número de runs")
    plt.xlabel("Vth(out1) [mV]")
    plt.ylabel("Vth(out0) [mV]")
    plt.title("TH_04 - Densidade de Vth(out0) vs Vth(out1)")
    plt.grid(True, alpha=0.35)
    plt.tight_layout()
    plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import re

# ============================================================
# CONFIGURAÇÃO DO TESTE — ADC SÍNCRONO
# ============================================================

FIN = 9e6              # frequência da senoide de entrada [Hz]
FS = 288e6             # frequência REAL de amostragem do ADC [Hz]

VOUT_THRESHOLD = 0.4   # threshold para converter outputs dos comparadores em bits

V_MIN = 0.0
V_MAX = 0.8

N_BITS = 4
N_CODES = 2**N_BITS
N_COMPARATORS = N_CODES - 1

LSB = (V_MAX - V_MIN) / N_CODES

# Ignorar transiente inicial.
# Exemplo: se quiseres ignorar os primeiros 5 us, usa 5e-6.
T_START_IGNORE = 0.0

# Fase de amostragem dentro do período de sampling.
# 0.0 -> amostra em t0 + nTs
# 0.5 -> amostra no meio do período de clock
# Ajusta se quiseres testar sensibilidade ao ponto de amostragem.
SAMPLE_PHASE = 0.0

# Número máximo de harmónicos usados para THD/SNR
N_HARMONICS = 7

# Janela temporal mostrada nos plots
PLOT_WINDOW_US = 1.0

# ============================================================
# PARÂMETROS DE COERÊNCIA FFT
# ============================================================

# Como FIN = 9 MHz e FS = 288 MHz:
# FS / FIN = 32.
#
# Isto significa que a senoide tem exatamente 32 amostras por período.
# Para uma FFT coerente, o número total de amostras N deve conter um
# número inteiro de períodos da senoide.
#
# Como cada período tem 32 amostras, basta garantir que:
#
#       N é múltiplo de 32
#
# O script faz isso automaticamente mais abaixo, truncando o vetor
# para o maior múltiplo de 32 possível.
#
# Exemplo:
#   N = 32768 amostras
#   N / 32 = 1024 períodos completos
#
# Tempo correspondente:
#   32768 / 288e6 = 113.78 us
#
# Recomenda-se simular pelo menos 120 us para ter margem.

SAMPLES_PER_PERIOD = int(round(FS / FIN))

print("===================================")
print("CONFIGURAÇÃO")
print("===================================")
print(f"FIN                 = {FIN/1e6:.3f} MHz")
print(f"FS ADC              = {FS/1e6:.3f} MHz")
print(f"FS/FIN              = {FS/FIN:.3f}")
print(f"Amostras por período= {SAMPLES_PER_PERIOD}")
print(f"V_MIN               = {V_MIN:.3f} V")
print(f"V_MAX               = {V_MAX:.3f} V")
print(f"LSB                 = {LSB*1e3:.3f} mV")
print(f"VOUT_THRESHOLD      = {VOUT_THRESHOLD:.3f} V")
print(f"SAMPLE_PHASE        = {SAMPLE_PHASE:.3f} Ts")


# ============================================================
# FUNÇÕES
# ============================================================

def read_waveview_file(filename):
    """
    Lê ficheiro WaveView.

    Usa apenas:
        v(/out<0>) ... v(/out<14>)

    Ignora:
        v(out<0>) ... v(out<14>)

    Isto é importante porque disseste que os outputs fornecidos são
    diretamente os 15 comparadores, sem encoder.
    """

    header = None
    header_line_idx = None

    with open(filename, "r") as f:
        for i, line in enumerate(f):
            s = line.strip()

            if s.startswith("XVAL"):
                header = s.split()
                header_line_idx = i
                break

    if header is None:
        raise ValueError("Header XVAL não encontrado.")

    out_dict = {}
    ignored_analog = []
    duplicate_out_info = []
    vin_col = None

    for i, name in enumerate(header):

        clean = (
            str(name)
            .replace("&amp;lt;", "<")
            .replace("&amp;gt;", ">")
            .replace("&lt;", "<")
            .replace("&gt;", ">")
            .strip()
            .lower()
        )

        # Tenta encontrar Vin.
        # Dependendo do export, pode aparecer como v(/vin+), v(vin+), v(/vin), etc.
        if (
            "v(/vin+" in clean
            or "v(vin+" in clean
            or "v(/vin)" in clean
            or "v(vin)" in clean
            or "vin" in clean and vin_col is None
        ):
            vin_col = i

        # Outputs digitais dos comparadores: v(/out<k>)
        m_digital = re.search(r"v\(/out<(\d+)>\)", clean)

        if m_digital:
            idx = int(m_digital.group(1))

            if 0 <= idx <= 14:
                if idx not in out_dict:
                    out_dict[idx] = (i, idx, name)
                else:
                    duplicate_out_info.append((i, idx, name))

            continue

        # Outputs analógicos v(out<k>) ignorados
        m_analog = re.search(r"v\(out<(\d+)>\)", clean)

        if m_analog:
            ignored_analog.append((i, int(m_analog.group(1)), name))

    if vin_col is None:
        raise ValueError("Coluna Vin não encontrada.")

    if len(out_dict) == 0:
        raise ValueError("Nenhuma coluna digital v(/out<0>)...v(/out<14>) encontrada.")

    # Ordenar por out<0>, out<1>, ..., out<14>
    out_info_sorted = [out_dict[k] for k in sorted(out_dict.keys())]

    out_cols = [x[0] for x in out_info_sorted]
    out_indices = [x[1] for x in out_info_sorted]

    print()
    print("===================================")
    print("COLUNAS USADAS")
    print("===================================")
    print(f"Vin column                   = {vin_col}")
    print(f"Outputs digitais usados      = {out_indices}")
    print(f"Colunas digitais usadas      = {out_cols}")
    print(f"N outputs digitais           = {len(out_indices)}")

    if len(ignored_analog) > 0:
        ignored_indices = [x[1] for x in ignored_analog]
        print(f"Outputs analógicos ignorados = {ignored_indices}")

    if len(duplicate_out_info) > 0:
        duplicate_indices = [x[1] for x in duplicate_out_info]
        print(f"Outputs digitais duplicados ignorados = {duplicate_indices}")

    if len(out_indices) != N_COMPARATORS:
        print()
        print("AVISO:")
        print(f"Foram encontrados {len(out_indices)} outputs digitais.")
        print(f"Esperado: {N_COMPARATORS}")

    df = pd.read_csv(
        filename,
        sep=r"\s+",
        comment="#",
        skiprows=header_line_idx + 1,
        header=None,
        engine="c"
    )

    data = df.to_numpy(dtype=float)

    t = data[:, 0]
    vin = data[:, vin_col]
    outs = data[:, out_cols]

    # Ordenar por tempo
    idx_sort = np.argsort(t)

    t = t[idx_sort]
    vin = vin[idx_sort]
    outs = outs[idx_sort, :]

    # Remover tempos duplicados
    t_unique, unique_idx = np.unique(t, return_index=True)

    t = t_unique
    vin = vin[unique_idx]
    outs = outs[unique_idx, :]

    return t, vin, outs, out_indices


def alias_frequency(f, fs):
    """
    Calcula frequência aliased para dentro de [0, fs/2].
    Útil para identificar harmónicos acima de Nyquist.
    """

    f_mod = np.mod(f, fs)

    if f_mod > fs / 2:
        f_alias = fs - f_mod
    else:
        f_alias = f_mod

    return f_alias


def calculate_fft_metrics(x, fs, fin, n_harmonics=7):
    """
    Calcula SINAD, ENOB, SNR, THD e SFDR.

    Nota:
    Se o número de amostras for coerente com FIN e FS, usa janela retangular.
    Caso contrário, usa Hann para reduzir leakage.
    """

    x = np.asarray(x)

    # Remover DC
    x = x - np.mean(x)

    N = len(x)

    fund_bin_float = fin / fs * N
    fund_bin = int(round(fund_bin_float))

    coherent = abs(fund_bin_float - fund_bin) < 1e-9

    print()
    print("===================================")
    print("COERÊNCIA FFT")
    print("===================================")
    print(f"N amostras FFT          = {N}")
    print(f"Bin fundamental ideal   = {fund_bin_float:.6f}")
    print(f"Bin fundamental usado   = {fund_bin}")
    print(f"Coerente                = {coherent}")

    if coherent:
        window = np.ones(N)
    else:
        print("AVISO: FFT não coerente. A usar janela Hann.")
        window = np.hanning(N)

    xw = x * window

    X = np.fft.rfft(xw)
    P = np.abs(X)**2
    P = P / np.sum(window**2)

    # Remover DC
    P[0] = 0

    freq = np.fft.rfftfreq(N, d=1/fs)

    if fund_bin <= 0 or fund_bin >= len(P):
        fund_bin = np.argmax(P[1:]) + 1

    # Bins da fundamental
    if coherent:
        signal_bins = [fund_bin]
    else:
        signal_bins = [fund_bin - 1, fund_bin, fund_bin + 1]
        signal_bins = [b for b in signal_bins if 0 <= b < len(P)]

    signal_power = np.sum(P[signal_bins])

    # ========================================================
    # Harmónicos
    # ========================================================

    harmonic_bins = []

    for h in range(2, n_harmonics + 1):

        f_h = alias_frequency(h * fin, fs)
        hb = int(round(f_h / (fs / N)))

        if hb <= 0 or hb >= len(P):
            continue

        # Evitar contar a fundamental como harmónico por alias
        if hb in signal_bins:
            continue

        if hb not in harmonic_bins:
            harmonic_bins.append(hb)

    harmonic_power = np.sum(P[harmonic_bins]) if len(harmonic_bins) > 0 else 0.0

    # ========================================================
    # SINAD
    # Fundamental / todo o resto exceto DC
    # ========================================================

    noise_dist_mask = np.ones(len(P), dtype=bool)
    noise_dist_mask[0] = False

    for b in signal_bins:
        noise_dist_mask[b] = False

    noise_dist_power = np.sum(P[noise_dist_mask])

    SINAD = 10 * np.log10(signal_power / noise_dist_power)
    ENOB = (SINAD - 1.76) / 6.02

    # ========================================================
    # SNR
    # Fundamental / ruído, excluindo harmónicos
    # ========================================================

    noise_only_mask = np.ones(len(P), dtype=bool)
    noise_only_mask[0] = False

    for b in signal_bins:
        noise_only_mask[b] = False

    for b in harmonic_bins:
        noise_only_mask[b] = False

    noise_only_power = np.sum(P[noise_only_mask])

    SNR = 10 * np.log10(signal_power / noise_only_power) if noise_only_power > 0 else np.inf

    # ========================================================
    # THD
    # Harmónicos / fundamental
    # ========================================================

    THD = 10 * np.log10(harmonic_power / signal_power) if harmonic_power > 0 else -np.inf

    # ========================================================
    # SFDR
    # Fundamental / maior spur
    # Exclui DC e fundamental
    # ========================================================

    spur_mask = np.ones(len(P), dtype=bool)
    spur_mask[0] = False

    for b in signal_bins:
        spur_mask[b] = False

    spur_power = np.max(P[spur_mask])
    spur_bin = np.argmax(np.where(spur_mask, P, -np.inf))

    SFDR = 10 * np.log10(signal_power / spur_power)

    return {
        "N": N,
        "freq": freq,
        "P": P,
        "fund_bin": fund_bin,
        "fund_freq": freq[fund_bin],
        "signal_bins": signal_bins,
        "harmonic_bins": harmonic_bins,
        "SINAD": SINAD,
        "ENOB": ENOB,
        "SNR": SNR,
        "THD": THD,
        "SFDR": SFDR,
        "spur_bin": spur_bin,
        "spur_freq": freq[spur_bin],
        "coherent": coherent
    }


def calculate_bubble_rate(bits):
    """
    Calcula bubble error rate para código termómetro.

    bits: matriz [N_amostras x N_comparadores]
          Cada linha contém os bits termómetro ordenados por out<0>...out<14>.

    Um padrão ideal de termómetro tem no máximo uma transição:
        111110000000000
    ou:
        000001111111111

    Bubble error ocorre se houver mais de uma transição entre 0 e 1.
    """

    bits = np.asarray(bits).astype(int)

    n_samples = bits.shape[0]

    bubble_flags = []
    n_transitions_list = []
    bubble_strength_list = []

    for row in bits:

        diff = np.diff(row)

        transitions = np.sum(diff != 0)

        has_bubble = transitions > 1

        bubble_flags.append(has_bubble)
        n_transitions_list.append(transitions)

        bubble_strength_list.append(max(0, transitions - 1))

    bubble_flags = np.array(bubble_flags)
    n_transitions_list = np.array(n_transitions_list)
    bubble_strength_list = np.array(bubble_strength_list)

    n_bubble_samples = np.sum(bubble_flags)

    bubble_rate = n_bubble_samples / n_samples
    bubble_rate_percent = 100 * bubble_rate

    mean_bubble_strength = np.mean(bubble_strength_list)
    max_bubble_strength = np.max(bubble_strength_list)

    return {
        "n_samples": n_samples,
        "n_bubble_samples": n_bubble_samples,
        "bubble_rate": bubble_rate,
        "bubble_rate_percent": bubble_rate_percent,
        "mean_bubble_strength": mean_bubble_strength,
        "max_bubble_strength": max_bubble_strength,
        "bubble_flags": bubble_flags,
        "n_transitions": n_transitions_list,
        "bubble_strength": bubble_strength_list
    }


def thermometer_to_code(bits):
    """
    Converte os 15 outputs dos comparadores em código ADC.

    Para um flash ADC ideal de 4 bits:
        número de comparadores = 2^4 - 1 = 15

    O código digital pode ser obtido contando o número de comparadores a '1':
        código = soma dos bits

    Resultado:
        código entre 0 e 15
    """

    adc_code = np.sum(bits, axis=1)

    adc_code = np.clip(adc_code, 0, N_COMPARATORS)

    return adc_code


def reconstruct_adc_voltage_from_code(adc_code):
    """
    Reconstrói a tensão equivalente a partir do código ADC.

    Como tens um ADC de 4 bits com V_MIN = 0 V e V_MAX = 0.8 V:

        LSB = (V_MAX - V_MIN) / 16 = 50 mV

    Existem 16 códigos: 0, 1, ..., 15.

    Uma reconstrução mais correta para análise é usar o centro de cada bin:

        Vrec = V_MIN + (code + 0.5) * LSB

    Assim:
        code 0  -> 25 mV
        code 1  -> 75 mV
        ...
        code 15 -> 775 mV

    Isto representa melhor o nível quantizado do ADC do que mapear
    diretamente code 0 para 0 V e code 15 para 0.8 V.
    """

    adc_code = np.asarray(adc_code)

    vrec = V_MIN + (adc_code + 0.5) * LSB

    # Garantir que não ultrapassa os centros válidos dos códigos
    vrec = np.clip(vrec, V_MIN + 0.5*LSB, V_MAX - 0.5*LSB)

    return vrec


# ============================================================
# UPLOAD DO FICHEIRO
# ============================================================

print()
print("===================================")
print("UPLOAD DO FICHEIRO")
print("===================================")

uploaded = files.upload()
filename = list(uploaded.keys())[0]

print(f"Ficheiro carregado: {filename}")


# ============================================================
# LER DADOS
# ============================================================

t_raw, vin_raw, outs_raw, out_indices = read_waveview_file(filename)

print()
print("===================================")
print("DADOS LIDOS")
print("===================================")
print(f"N pontos brutos    = {len(t_raw)}")
print(f"Tempo inicial      = {t_raw[0]*1e6:.3f} us")
print(f"Tempo final        = {t_raw[-1]*1e6:.3f} us")
print(f"Tempo total        = {(t_raw[-1] - t_raw[0])*1e6:.3f} us")


# ============================================================
# CORTAR TRANSIENTE INICIAL
# ============================================================

mask_time = t_raw >= (t_raw[0] + T_START_IGNORE)

t_raw = t_raw[mask_time]
vin_raw = vin_raw[mask_time]
outs_raw = outs_raw[mask_time, :]

print()
print("===================================")
print("APÓS REMOVER TRANSIENTE")
print("===================================")
print(f"T_START_IGNORE     = {T_START_IGNORE*1e6:.3f} us")
print(f"N pontos restantes = {len(t_raw)}")
print(f"Tempo inicial útil = {t_raw[0]*1e6:.3f} us")
print(f"Tempo final útil   = {t_raw[-1]*1e6:.3f} us")


# ============================================================
# CRIAR INSTANTES DE AMOSTRAGEM SÍNCRONOS
# ============================================================

TS = 1 / FS

t_start = t_raw[0]
t_stop = t_raw[-1]

# Primeiro instante de amostragem
t_first_sample = t_start + SAMPLE_PHASE * TS

# Grelha síncrona de amostragem
t_sample = np.arange(t_first_sample, t_stop, TS)

if len(t_sample) < 100:
    raise ValueError("Poucas amostras síncronas. Aumenta o tempo de simulação.")

print()
print("===================================")
print("AMOSTRAGEM SÍNCRONA")
print("===================================")
print(f"FS ADC             = {FS/1e6:.3f} MHz")
print(f"TS ADC             = {TS*1e9:.6f} ns")
print(f"N amostras inicial = {len(t_sample)}")
print(f"Tempo útil inicial = {(t_sample[-1]-t_sample[0])*1e6:.3f} us")
print(f"FIN                = {FIN/1e6:.3f} MHz")
print(f"FS/FIN             = {FS/FIN:.3f}")
print(f"Amostras por ciclo = {SAMPLES_PER_PERIOD}")


# ============================================================
# AMOSTRAR VIN E OUTPUTS DOS COMPARADORES
# ============================================================

# Vin é analógico, por isso interpolamos apenas para visualizar e comparar.
vin_sampled = np.interp(t_sample, t_raw, vin_raw)

# Os outputs dos comparadores podem vir com tempos não exatamente alinhados
# com os instantes de sampling. Por isso interpolamos a waveform e depois
# aplicamos threshold no instante de sampling.
outs_sampled = np.zeros((len(t_sample), outs_raw.shape[1]))

for k in range(outs_raw.shape[1]):
    outs_sampled[:, k] = np.interp(t_sample, t_raw, outs_raw[:, k])


# ============================================================
# CONVERTER OUTPUTS PARA BITS
# ============================================================

bits = (outs_sampled > VOUT_THRESHOLD).astype(int)


# ============================================================
# CÓDIGO ADC A PARTIR DOS 15 COMPARADORES
# ============================================================

adc_code = thermometer_to_code(bits)


# ============================================================
# RECONSTRUIR VALOR ANALÓGICO EQUIVALENTE
# ============================================================

adc_reconstructed = reconstruct_adc_voltage_from_code(adc_code)


# ============================================================
# TRUNCAMENTO PARA FFT COERENTE
# ============================================================

# Com FIN = 9 MHz e FS = 288 MHz:
#
#     FS / FIN = 32 amostras por período
#
# Para uma FFT coerente, o bloco analisado deve conter um número inteiro
# de períodos da senoide. Como cada período tem 32 amostras, fazemos:
#
#     N_COHERENT = maior múltiplo de 32 menor ou igual ao número disponível
#
# Isto evita leakage espectral sem precisar de janela Hann.
#
# Exemplo:
#     Se houver 34567 amostras:
#     N_COHERENT = floor(34567 / 32) * 32 = 34560
#
# Assim a FFT fica alinhada com a senoide de entrada.

N_AVAILABLE = len(adc_reconstructed)
N_COHERENT = (N_AVAILABLE // SAMPLES_PER_PERIOD) * SAMPLES_PER_PERIOD

if N_COHERENT < SAMPLES_PER_PERIOD * 10:
    raise ValueError("Poucas amostras coerentes para FFT. Aumenta o tempo de simulação.")

# Aplicar truncamento a todos os vetores síncronos
t_sample = t_sample[:N_COHERENT]
vin_sampled = vin_sampled[:N_COHERENT]
outs_sampled = outs_sampled[:N_COHERENT, :]
bits = bits[:N_COHERENT, :]
adc_code = adc_code[:N_COHERENT]
adc_reconstructed = adc_reconstructed[:N_COHERENT]

N_PERIODS_USED = N_COHERENT // SAMPLES_PER_PERIOD

print()
print("===================================")
print("TRUNCAMENTO PARA FFT COERENTE")
print("===================================")
print(f"Amostras disponíveis       = {N_AVAILABLE}")
print(f"Amostras usadas na FFT     = {N_COHERENT}")
print(f"Amostras removidas         = {N_AVAILABLE - N_COHERENT}")
print(f"Amostras por período       = {SAMPLES_PER_PERIOD}")
print(f"Períodos inteiros usados   = {N_PERIODS_USED}")
print(f"Tempo usado na FFT         = {N_COHERENT/FS*1e6:.3f} us")
print(f"Resolução da FFT           = {FS/N_COHERENT/1e3:.3f} kHz")


# ============================================================
# BUBBLE ERROR RATE
# ============================================================

bubble = calculate_bubble_rate(bits)


# ============================================================
# CÁLCULO FFT / SINAD / ENOB / SNR / THD / SFDR
# ============================================================

metrics = calculate_fft_metrics(
    x=adc_reconstructed,
    fs=FS,
    fin=FIN,
    n_harmonics=N_HARMONICS
)


# ============================================================
# RESULTADOS
# ============================================================

print()
print("===================================")
print("RESULTADOS DINÂMICOS")
print("===================================")
print(f"fin                  = {FIN/1e6:.3f} MHz")
print(f"FS ADC               = {FS/1e6:.3f} MHz")
print(f"N FFT                = {metrics['N']}")
print(f"Fundamental bin      = {metrics['fund_bin']}")
print(f"Fundamental freq     = {metrics['fund_freq']/1e6:.6f} MHz")
print(f"SINAD                = {metrics['SINAD']:.2f} dB")
print(f"ENOB                 = {metrics['ENOB']:.2f} bits")
print(f"SNR                  = {metrics['SNR']:.2f} dB")
print(f"THD                  = {metrics['THD']:.2f} dB")
print(f"SFDR                 = {metrics['SFDR']:.2f} dB")
print(f"Maior spur           = {metrics['spur_freq']/1e6:.6f} MHz")
print(f"Código mínimo        = {np.min(adc_code)}")
print(f"Código máximo        = {np.max(adc_code)}")
print(f"Códigos usados       = {np.unique(adc_code)}")

print()
print("===================================")
print("BUBBLE ERRORS")
print("===================================")
print(f"Amostras analisadas       = {bubble['n_samples']}")
print(f"Amostras com bubble       = {bubble['n_bubble_samples']}")
print(f"Bubble rate               = {bubble['bubble_rate_percent']:.4f} %")
print(f"Bubble strength médio     = {bubble['mean_bubble_strength']:.4f}")
print(f"Bubble strength máximo    = {bubble['max_bubble_strength']}")


# ============================================================
# SENOIDE / SLEW RATE
# ============================================================

VIN_AMPLITUDE_EST = (np.max(vin_sampled) - np.min(vin_sampled)) / 2
VIN_OFFSET_EST = (np.max(vin_sampled) + np.min(vin_sampled)) / 2

SR_THEORETICAL = 2 * np.pi * FIN * VIN_AMPLITUDE_EST
SR_THEORETICAL_V_PER_US = SR_THEORETICAL / 1e6

print()
print("===================================")
print("SENOIDE / SLEW RATE")
print("===================================")
print(f"Amplitude estimada   = {VIN_AMPLITUDE_EST:.3f} V")
print(f"Offset estimado      = {VIN_OFFSET_EST:.3f} V")
print(f"SR máximo estimado   = {SR_THEORETICAL_V_PER_US:.3f} V/us")


# ============================================================
# PLOTS TEMPORAIS
# ============================================================

t_us = t_sample * 1e6
t0_us = t_us[0]
t1_us = t0_us + PLOT_WINDOW_US

plot_mask = (t_us >= t0_us) & (t_us <= t1_us)

plt.figure(figsize=(12,4))
plt.plot(t_us[plot_mask], vin_sampled[plot_mask], label="Vin")
plt.step(
    t_us[plot_mask],
    adc_reconstructed[plot_mask],
    where="post",
    label="ADC reconstruído",
    alpha=0.8
)
plt.title("Vin vs ADC reconstruído — ADC síncrono")
plt.xlabel("Tempo [µs]")
plt.ylabel("Tensão [V]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,4))
plt.step(
    t_us[plot_mask],
    adc_code[plot_mask],
    where="post"
)
plt.plot(
    t_us[plot_mask],
    adc_code[plot_mask],
    marker='.',
    linestyle='None',
    markersize=3
)
plt.title("Código ADC reconstruído — ADC síncrono")
plt.xlabel("Tempo [µs]")
plt.ylabel("Código")
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,4))
plt.step(
    t_us[plot_mask],
    bubble["bubble_flags"][plot_mask].astype(int),
    where="post"
)
plt.plot(
    t_us[plot_mask],
    bubble["bubble_flags"][plot_mask].astype(int),
    marker='.',
    linestyle='None',
    markersize=3
)
plt.title("Bubble errors — ADC síncrono")
plt.xlabel("Tempo [µs]")
plt.ylabel("Bubble error")
plt.yticks([0, 1], ["No bubble", "Bubble"])
plt.grid(True)
plt.tight_layout()
plt.show()


# ============================================================
# PLOT FFT
# ============================================================

plt.figure(figsize=(12,5))
plt.plot(metrics["freq"] / 1e6, 10*np.log10(metrics["P"] + 1e-30))
plt.axvline(FIN/1e6, color="red", linestyle="--", label="Fundamental")
plt.axvline(metrics["spur_freq"]/1e6, color="orange", linestyle="--", label="Maior spur")
plt.title("FFT do ADC reconstruído — ADC síncrono")
plt.xlabel("Frequência [MHz]")
plt.ylabel("Potência [dB]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# HISTOGRAMA DOS CÓDIGOS
# ============================================================

plt.figure(figsize=(8,4))
plt.hist(
    adc_code,
    bins=np.arange(-0.5, N_COMPARATORS + 1.5, 1),
    edgecolor="black"
)
plt.title("Histograma dos códigos ADC")
plt.xlabel("Código")
plt.ylabel("Contagem")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import re

# ============================================================
# TEST CONFIGURATION — SYNCHRONOUS ADC
# ============================================================

FIN = 9e6
FS = 288e6

VOUT_THRESHOLD = 0.4

V_MIN = 0.0
V_MAX = 0.8

N_BITS = 4
N_CODES = 2**N_BITS
N_COMPARATORS = N_CODES - 1

LSB = (V_MAX - V_MIN) / N_CODES

# For 20 us simulation, usually keep this at 0.
# If needed, use for example 0.5e-6 or 1e-6.
T_START_IGNORE = 0.0

# Main sampling phase used for final analysis.
# 0.0 -> sample at t0 + nTs
# 0.5 -> sample halfway through the ADC sampling period
SAMPLE_PHASE = 0.5

# Optional sampling phase sweep.
DO_PHASE_SWEEP = True

# Sampling phases to test, expressed as a fraction of TS.
PHASE_LIST = [0.0, 0.25, 0.5, 0.75]

# If True, the script automatically selects the phase with the best SINAD.
AUTO_SELECT_BEST_PHASE = False

N_HARMONICS = 7

PLOT_WINDOW_US = 1.0

EXPORT_CSV = True

SAMPLES_PER_PERIOD = int(round(FS / FIN))

# ============================================================
# RAW WAVEFORM CODE PRESENCE
# ============================================================

# This checks which codes exist anywhere in the exported waveform.
# This is different from the codes captured only at ADC sampling instants.
MEASURE_RAW_CODE_PRESENCE = True

# Minimum accumulated duration for a code to be considered present.
# With 1 ns exported step:
#   0.0   -> any occurrence counts
#   1e-9  -> code must last at least ~1 ns
#   2e-9  -> code must last at least ~2 ns
MIN_CODE_OBS_TIME = 0.0

print("===================================")
print("CONFIGURATION")
print("===================================")
print(f"FIN                  = {FIN/1e6:.3f} MHz")
print(f"FS ADC               = {FS/1e6:.3f} MHz")
print(f"FS/FIN               = {FS/FIN:.3f}")
print(f"Samples per period   = {SAMPLES_PER_PERIOD}")
print(f"V_MIN                = {V_MIN:.3f} V")
print(f"V_MAX                = {V_MAX:.3f} V")
print(f"LSB                  = {LSB*1e3:.3f} mV")
print(f"VOUT_THRESHOLD       = {VOUT_THRESHOLD:.3f} V")
print(f"SAMPLE_PHASE         = {SAMPLE_PHASE:.3f} Ts")


# ============================================================
# FUNCTIONS
# ============================================================

def clean_header_name(name):
    return (
        str(name)
        .replace("&amp;amp;lt;", "<")
        .replace("&amp;amp;gt;", ">")
        .replace("&amp;lt;", "<")
        .replace("&amp;gt;", ">")
        .replace("&lt;", "<")
        .replace("&gt;", ">")
        .strip()
        .lower()
    )


def read_waveview_file(filename):
    """
    Reads a WaveView exported file.

    Used columns:
        v(/out<0>) ... v(/out<14>)

    Ignored columns:
        v(out<0>) ... v(out<14>)

    The selected outputs are the 15 comparator outputs before encoding.
    """

    header = None
    header_line_idx = None

    with open(filename, "r") as f:
        for i, line in enumerate(f):
            s = line.strip()
            if s.startswith("XVAL"):
                header = s.split()
                header_line_idx = i
                break

    if header is None:
        raise ValueError("Header XVAL not found.")

    out_dict = {}
    ignored_analog = []
    duplicate_out_info = []
    vin_col = None

    for i, name in enumerate(header):

        clean = clean_header_name(name)

        # Flexible Vin detection.
        if (
            ("v(/vin+" in clean)
            or ("v(vin+" in clean)
            or ("v(/vin)" in clean)
            or ("v(vin)" in clean)
            or (("vin" in clean) and (vin_col is None))
        ):
            vin_col = i

        # Digital comparator outputs: v(/out<k>)
        m_digital = re.search(r"v\(/out<(\d+)>\)", clean)

        if m_digital:
            idx = int(m_digital.group(1))

            if 0 <= idx <= 14:
                if idx not in out_dict:
                    out_dict[idx] = (i, idx, name)
                else:
                    duplicate_out_info.append((i, idx, name))

            continue

        # Analog outputs v(out<k>) are ignored.
        m_analog = re.search(r"v\(out<(\d+)>\)", clean)

        if m_analog:
            ignored_analog.append((i, int(m_analog.group(1)), name))

    if vin_col is None:
        raise ValueError("Vin column not found.")

    if len(out_dict) == 0:
        raise ValueError("No digital columns v(/out<0>)...v(/out<14>) found.")

    out_info_sorted = [out_dict[k] for k in sorted(out_dict.keys())]

    out_cols = [x[0] for x in out_info_sorted]
    out_indices = [x[1] for x in out_info_sorted]

    print()
    print("===================================")
    print("USED COLUMNS")
    print("===================================")
    print(f"Vin column                   = {vin_col}")
    print(f"Digital outputs used         = {out_indices}")
    print(f"Digital columns used         = {out_cols}")
    print(f"Number of digital outputs    = {len(out_indices)}")

    if len(ignored_analog) > 0:
        ignored_indices = [x[1] for x in ignored_analog]
        print(f"Analog outputs ignored       = {ignored_indices}")

    if len(duplicate_out_info) > 0:
        duplicate_indices = [x[1] for x in duplicate_out_info]
        print(f"Duplicate digital outputs ignored = {duplicate_indices}")

    if len(out_indices) != N_COMPARATORS:
        print()
        print("WARNING:")
        print(f"Found {len(out_indices)} digital outputs.")
        print(f"Expected {N_COMPARATORS}.")

    df = pd.read_csv(
        filename,
        sep=r"\s+",
        comment="#",
        skiprows=header_line_idx + 1,
        header=None,
        engine="c"
    )

    data = df.to_numpy(dtype=float)

    t = data[:, 0]
    vin = data[:, vin_col]
    outs = data[:, out_cols]

    # Sort by time.
    idx_sort = np.argsort(t)

    t = t[idx_sort]
    vin = vin[idx_sort]
    outs = outs[idx_sort, :]

    # Remove duplicate time points.
    t_unique, unique_idx = np.unique(t, return_index=True)

    t = t_unique
    vin = vin[unique_idx]
    outs = outs[unique_idx, :]

    return t, vin, outs, out_indices


def alias_frequency(f, fs):
    """
    Folds a frequency into the [0, fs/2] Nyquist band.
    """

    f_mod = np.mod(f, fs)

    if f_mod > fs / 2:
        return fs - f_mod
    else:
        return f_mod


def calculate_fft_metrics(x, fs, fin, n_harmonics=7, verbose=True):
    """
    Calculates SINAD, ENOB, SNR, THD and SFDR.
    """

    x = np.asarray(x)
    x = x - np.mean(x)

    N = len(x)

    fund_bin_float = fin / fs * N
    fund_bin = int(round(fund_bin_float))

    coherent = abs(fund_bin_float - fund_bin) < 1e-9

    if verbose:
        print()
        print("===================================")
        print("FFT COHERENCE")
        print("===================================")
        print(f"N FFT samples           = {N}")
        print(f"Ideal fundamental bin   = {fund_bin_float:.6f}")
        print(f"Used fundamental bin    = {fund_bin}")
        print(f"Coherent                = {coherent}")

    if coherent:
        window = np.ones(N)
    else:
        if verbose:
            print("WARNING: Non-coherent FFT. Hann window is used.")
        window = np.hanning(N)

    xw = x * window

    X = np.fft.rfft(xw)
    P = np.abs(X)**2
    P = P / np.sum(window**2)

    P[0] = 0.0

    freq = np.fft.rfftfreq(N, d=1/fs)

    if fund_bin <= 0 or fund_bin >= len(P):
        fund_bin = np.argmax(P[1:]) + 1

    if coherent:
        signal_bins = [fund_bin]
    else:
        signal_bins = [fund_bin - 1, fund_bin, fund_bin + 1]
        signal_bins = [b for b in signal_bins if 0 <= b < len(P)]

    signal_power = np.sum(P[signal_bins])

    harmonic_bins = []

    for h in range(2, n_harmonics + 1):

        f_h = alias_frequency(h * fin, fs)
        hb = int(round(f_h / (fs / N)))

        if hb <= 0 or hb >= len(P):
            continue

        if hb in signal_bins:
            continue

        if hb not in harmonic_bins:
            harmonic_bins.append(hb)

    harmonic_power = np.sum(P[harmonic_bins]) if len(harmonic_bins) > 0 else 0.0

    # SINAD: signal / all non-signal non-DC power.
    noise_dist_mask = np.ones(len(P), dtype=bool)
    noise_dist_mask[0] = False

    for b in signal_bins:
        noise_dist_mask[b] = False

    noise_dist_power = np.sum(P[noise_dist_mask])

    SINAD = 10 * np.log10(signal_power / noise_dist_power)
    ENOB = (SINAD - 1.76) / 6.02

    # SNR: signal / noise excluding harmonics.
    noise_only_mask = np.ones(len(P), dtype=bool)
    noise_only_mask[0] = False

    for b in signal_bins:
        noise_only_mask[b] = False

    for b in harmonic_bins:
        noise_only_mask[b] = False

    noise_only_power = np.sum(P[noise_only_mask])

    SNR = 10 * np.log10(signal_power / noise_only_power) if noise_only_power > 0 else np.inf

    # THD: harmonics / signal.
    THD = 10 * np.log10(harmonic_power / signal_power) if harmonic_power > 0 else -np.inf

    # SFDR: signal / largest spur.
    spur_mask = np.ones(len(P), dtype=bool)
    spur_mask[0] = False

    for b in signal_bins:
        spur_mask[b] = False

    spur_power = np.max(P[spur_mask])
    spur_bin = np.argmax(np.where(spur_mask, P, -np.inf))

    SFDR = 10 * np.log10(signal_power / spur_power)

    return {
        "N": N,
        "freq": freq,
        "P": P,
        "fund_bin": fund_bin,
        "fund_freq": freq[fund_bin],
        "signal_bins": signal_bins,
        "harmonic_bins": harmonic_bins,
        "SINAD": SINAD,
        "ENOB": ENOB,
        "SNR": SNR,
        "THD": THD,
        "SFDR": SFDR,
        "spur_bin": spur_bin,
        "spur_freq": freq[spur_bin],
        "coherent": coherent
    }


def calculate_bubble_rate(bits):
    """
    Calculates bubble error rate for thermometer code.

    A valid thermometer pattern has at most one transition:
        111110000000000
    or:
        000001111111111
    """

    bits = np.asarray(bits).astype(int)

    n_samples = bits.shape[0]

    bubble_flags = []
    n_transitions_list = []
    bubble_strength_list = []

    for row in bits:

        diff = np.diff(row)

        transitions = np.sum(diff != 0)
        has_bubble = transitions > 1

        bubble_flags.append(has_bubble)
        n_transitions_list.append(transitions)
        bubble_strength_list.append(max(0, transitions - 1))

    bubble_flags = np.array(bubble_flags)
    n_transitions_list = np.array(n_transitions_list)
    bubble_strength_list = np.array(bubble_strength_list)

    n_bubble_samples = np.sum(bubble_flags)

    bubble_rate = n_bubble_samples / n_samples
    bubble_rate_percent = 100 * bubble_rate

    mean_bubble_strength = np.mean(bubble_strength_list)
    max_bubble_strength = np.max(bubble_strength_list)

    return {
        "n_samples": n_samples,
        "n_bubble_samples": n_bubble_samples,
        "bubble_rate": bubble_rate,
        "bubble_rate_percent": bubble_rate_percent,
        "mean_bubble_strength": mean_bubble_strength,
        "max_bubble_strength": max_bubble_strength,
        "bubble_flags": bubble_flags,
        "n_transitions": n_transitions_list,
        "bubble_strength": bubble_strength_list
    }


def thermometer_to_code(bits, vin_sampled=None):
    """
    Converts the 15 comparator outputs into ADC code.

    Default:
        code = number of comparators at logic '1'

    If vin_sampled is given, the function checks the polarity.
    If the code decreases when Vin increases, it automatically inverts it.
    """

    code_sum = np.sum(bits, axis=1)
    code_sum = np.clip(code_sum, 0, N_COMPARATORS)

    polarity = "normal"

    if vin_sampled is not None:
        corr = np.corrcoef(vin_sampled, code_sum)[0, 1]

        if np.isnan(corr):
            corr = 1.0

        if corr < 0:
            adc_code = N_COMPARATORS - code_sum
            polarity = "inverted"
        else:
            adc_code = code_sum
            polarity = "normal"
    else:
        adc_code = code_sum

    adc_code = np.clip(adc_code, 0, N_COMPARATORS)

    return adc_code.astype(int), polarity


def measure_code_presence_raw_waveform(t_raw, vin_raw, outs_raw):
    """
    Measures which ADC codes are present in the raw exported waveform.

    This is different from checking codes only at the synchronous ADC
    sampling instants.

    Method:
        1. Threshold comparator outputs at every raw exported time point.
        2. Convert thermometer bits to ADC code.
        3. Estimate accumulated time spent in each code.
        4. A code is considered present if its accumulated duration is above
           MIN_CODE_OBS_TIME.

    This is only used for code-presence diagnostics.
    It is not used for FFT/SINAD/ENOB calculations.
    """

    t_raw = np.asarray(t_raw)
    vin_raw = np.asarray(vin_raw)
    outs_raw = np.asarray(outs_raw)

    bits_raw = (outs_raw > VOUT_THRESHOLD).astype(int)

    code_sum = np.sum(bits_raw, axis=1)
    code_sum = np.clip(code_sum, 0, N_COMPARATORS)

    # Detect polarity using correlation with Vin.
    corr = np.corrcoef(vin_raw, code_sum)[0, 1]

    if np.isnan(corr):
        corr = 1.0

    if corr < 0:
        code_raw = N_COMPARATORS - code_sum
        polarity_raw = "inverted"
    else:
        code_raw = code_sum
        polarity_raw = "normal"

    code_raw = np.clip(code_raw, 0, N_COMPARATORS).astype(int)

    # Count raw occurrences.
    code_counts = np.bincount(code_raw, minlength=N_CODES)

    # Estimate accumulated duration per code.
    # Each point represents the interval until the next point.
    dt = np.diff(t_raw)

    code_time = np.zeros(N_CODES)

    for i in range(len(dt)):
        code_time[code_raw[i]] += dt[i]

    observed_codes = np.where(code_time > MIN_CODE_OBS_TIME)[0]
    missing_codes = sorted(set(range(N_CODES)) - set(observed_codes))

    return {
        "bits_raw": bits_raw,
        "code_raw": code_raw,
        "code_counts": code_counts,
        "code_time": code_time,
        "observed_codes": observed_codes,
        "missing_codes": missing_codes,
        "polarity_raw": polarity_raw,
        "correlation_raw": corr
    }


def reconstruct_adc_voltage_from_code(adc_code):
    """
    Reconstructs the equivalent analog voltage using code-bin centers.

    Vrec = V_MIN + (code + 0.5) * LSB
    """

    adc_code = np.asarray(adc_code)

    vrec = V_MIN + (adc_code + 0.5) * LSB
    vrec = np.clip(vrec, V_MIN + 0.5*LSB, V_MAX - 0.5*LSB)

    return vrec


def ideal_adc_from_vin(vin):
    """
    Generates an ideal ADC code and reconstructed voltage from Vin.

    This is used only as a reference in plots and exported data.
    """

    vin_clip = np.clip(vin, V_MIN, V_MAX - 1e-15)

    ideal_code = np.floor((vin_clip - V_MIN) / LSB).astype(int)
    ideal_code = np.clip(ideal_code, 0, N_CODES - 1)

    ideal_vrec = reconstruct_adc_voltage_from_code(ideal_code)

    return ideal_code, ideal_vrec


def analyze_for_phase(t_raw, vin_raw, outs_raw, phase, verbose=True):
    """
    Performs the full ADC analysis for a given sampling phase.

    The raw file may have a 1 ns output step.
    This function interpolates Vin and comparator outputs to the real ADC
    sampling instants:
        t = t0 + phase*TS + n*TS
    """

    TS = 1 / FS

    t_start = t_raw[0]
    t_stop = t_raw[-1]

    t_first_sample = t_start + phase * TS

    t_sample = np.arange(t_first_sample, t_stop, TS)

    if len(t_sample) < 100:
        raise ValueError("Too few synchronous samples. Increase simulation time.")

    vin_sampled = np.interp(t_sample, t_raw, vin_raw)

    outs_sampled = np.zeros((len(t_sample), outs_raw.shape[1]))

    for k in range(outs_raw.shape[1]):
        outs_sampled[:, k] = np.interp(t_sample, t_raw, outs_raw[:, k])

    bits = (outs_sampled > VOUT_THRESHOLD).astype(int)

    adc_code, polarity = thermometer_to_code(bits, vin_sampled=vin_sampled)

    adc_reconstructed = reconstruct_adc_voltage_from_code(adc_code)

    # Coherent FFT truncation.
    # Since FS/FIN = 32, the number of samples must be a multiple of 32.
    N_AVAILABLE = len(adc_reconstructed)
    N_COHERENT = (N_AVAILABLE // SAMPLES_PER_PERIOD) * SAMPLES_PER_PERIOD

    if N_COHERENT < SAMPLES_PER_PERIOD * 10:
        raise ValueError("Too few coherent samples for FFT. Increase simulation time.")

    t_sample = t_sample[:N_COHERENT]
    vin_sampled = vin_sampled[:N_COHERENT]
    outs_sampled = outs_sampled[:N_COHERENT, :]
    bits = bits[:N_COHERENT, :]
    adc_code = adc_code[:N_COHERENT]
    adc_reconstructed = adc_reconstructed[:N_COHERENT]

    ideal_code, ideal_reconstructed = ideal_adc_from_vin(vin_sampled)

    bubble = calculate_bubble_rate(bits)

    metrics = calculate_fft_metrics(
        x=adc_reconstructed,
        fs=FS,
        fin=FIN,
        n_harmonics=N_HARMONICS,
        verbose=verbose
    )

    if verbose:
        print()
        print("===================================")
        print("SYNCHRONOUS SAMPLING")
        print("===================================")
        print(f"Sampling phase          = {phase:.3f} Ts")
        print(f"FS ADC                  = {FS/1e6:.3f} MHz")
        print(f"TS ADC                  = {TS*1e9:.6f} ns")
        print(f"Available samples       = {N_AVAILABLE}")
        print(f"Samples used            = {N_COHERENT}")
        print(f"Removed samples         = {N_AVAILABLE - N_COHERENT}")
        print(f"Integer periods used    = {N_COHERENT // SAMPLES_PER_PERIOD}")
        print(f"FFT time window         = {N_COHERENT/FS*1e6:.3f} us")
        print(f"FFT resolution          = {FS/N_COHERENT/1e3:.3f} kHz")
        print(f"ADC polarity            = {polarity}")

    return {
        "phase": phase,
        "t_sample": t_sample,
        "vin_sampled": vin_sampled,
        "outs_sampled": outs_sampled,
        "bits": bits,
        "adc_code": adc_code,
        "adc_reconstructed": adc_reconstructed,
        "ideal_code": ideal_code,
        "ideal_reconstructed": ideal_reconstructed,
        "bubble": bubble,
        "metrics": metrics,
        "polarity": polarity
    }


# ============================================================
# FILE UPLOAD
# ============================================================

print()
print("===================================")
print("FILE UPLOAD")
print("===================================")

uploaded = files.upload()
filename = list(uploaded.keys())[0]

print(f"Loaded file: {filename}")


# ============================================================
# READ DATA
# ============================================================

t_raw, vin_raw, outs_raw, out_indices = read_waveview_file(filename)

print()
print("===================================")
print("RAW DATA")
print("===================================")
print(f"Raw points        = {len(t_raw)}")
print(f"Initial time      = {t_raw[0]*1e6:.3f} us")
print(f"Final time        = {t_raw[-1]*1e6:.3f} us")
print(f"Total time        = {(t_raw[-1] - t_raw[0])*1e6:.3f} us")

raw_dt = np.diff(t_raw)
print(f"Median raw step   = {np.median(raw_dt)*1e9:.3f} ns")
print(f"Min raw step      = {np.min(raw_dt)*1e9:.3f} ns")
print(f"Max raw step      = {np.max(raw_dt)*1e9:.3f} ns")

if np.median(raw_dt) > (1 / FS):
    print()
    print("WARNING:")
    print("Raw output step is larger than the ADC sampling period.")
    print("Interpolation will be used, but fast comparator transitions may not be fully captured.")


# ============================================================
# REMOVE INITIAL TRANSIENT
# ============================================================

mask_time = t_raw >= (t_raw[0] + T_START_IGNORE)

t_raw = t_raw[mask_time]
vin_raw = vin_raw[mask_time]
outs_raw = outs_raw[mask_time, :]

print()
print("===================================")
print("AFTER TRANSIENT REMOVAL")
print("===================================")
print(f"T_START_IGNORE     = {T_START_IGNORE*1e6:.3f} us")
print(f"Remaining points   = {len(t_raw)}")
print(f"Useful start time  = {t_raw[0]*1e6:.3f} us")
print(f"Useful final time  = {t_raw[-1]*1e6:.3f} us")


# ============================================================
# RAW WAVEFORM CODE PRESENCE
# ============================================================

if MEASURE_RAW_CODE_PRESENCE:

    raw_code_presence = measure_code_presence_raw_waveform(
        t_raw=t_raw,
        vin_raw=vin_raw,
        outs_raw=outs_raw
    )

    print()
    print("===================================")
    print("RAW WAVEFORM CODE PRESENCE")
    print("===================================")
    print(f"Raw code polarity        = {raw_code_presence['polarity_raw']}")
    print(f"Raw Vin/code correlation = {raw_code_presence['correlation_raw']:.6f}")
    print(f"Observed codes raw       = {raw_code_presence['observed_codes']}")
    print(f"Missing codes raw        = {raw_code_presence['missing_codes']}")

    print()
    print("Raw code accumulated time:")
    print("code | count | time [ns]")
    for k in range(N_CODES):
        print(
            f"{k:4d} | "
            f"{raw_code_presence['code_counts'][k]:5d} | "
            f"{raw_code_presence['code_time'][k]*1e9:9.3f}"
        )

else:
    raw_code_presence = None


# ============================================================
# OPTIONAL PHASE SWEEP
# ============================================================

phase_results = []

if DO_PHASE_SWEEP:

    print()
    print("===================================")
    print("SAMPLING PHASE SWEEP")
    print("===================================")
    print("phase | N    | SINAD [dB] | ENOB [bits] | SNR [dB] | THD [dB] | SFDR [dB] | min | max | not sampled codes | bubble [%]")

    for ph in PHASE_LIST:
        result_ph = analyze_for_phase(
            t_raw=t_raw,
            vin_raw=vin_raw,
            outs_raw=outs_raw,
            phase=ph,
            verbose=False
        )

        metrics_ph = result_ph["metrics"]
        bubble_ph = result_ph["bubble"]
        codes_used_ph = np.unique(result_ph["adc_code"])
        not_sampled_codes = sorted(set(range(N_CODES)) - set(codes_used_ph))

        phase_results.append({
            "phase": ph,
            "N": metrics_ph["N"],
            "SINAD": metrics_ph["SINAD"],
            "ENOB": metrics_ph["ENOB"],
            "SNR": metrics_ph["SNR"],
            "THD": metrics_ph["THD"],
            "SFDR": metrics_ph["SFDR"],
            "min_code": np.min(result_ph["adc_code"]),
            "max_code": np.max(result_ph["adc_code"]),
            "not_sampled_codes": not_sampled_codes,
            "bubble_rate_percent": bubble_ph["bubble_rate_percent"]
        })

        print(
            f"{ph:5.2f} | "
            f"{metrics_ph['N']:4d} | "
            f"{metrics_ph['SINAD']:10.2f} | "
            f"{metrics_ph['ENOB']:11.2f} | "
            f"{metrics_ph['SNR']:8.2f} | "
            f"{metrics_ph['THD']:8.2f} | "
            f"{metrics_ph['SFDR']:9.2f} | "
            f"{np.min(result_ph['adc_code']):3d} | "
            f"{np.max(result_ph['adc_code']):3d} | "
            f"{not_sampled_codes} | "
            f"{bubble_ph['bubble_rate_percent']:.4f}"
        )

    if AUTO_SELECT_BEST_PHASE:
        best = max(phase_results, key=lambda r: r["SINAD"])
        SAMPLE_PHASE = best["phase"]

        print()
        print("===================================")
        print("AUTO SELECTED PHASE")
        print("===================================")
        print(f"Selected phase = {SAMPLE_PHASE:.3f} Ts")
        print(f"Reason         = best SINAD in phase sweep")


# ============================================================
# FINAL ANALYSIS FOR SELECTED PHASE
# ============================================================

result = analyze_for_phase(
    t_raw=t_raw,
    vin_raw=vin_raw,
    outs_raw=outs_raw,
    phase=SAMPLE_PHASE,
    verbose=True
)

t_sample = result["t_sample"]
vin_sampled = result["vin_sampled"]
outs_sampled = result["outs_sampled"]
bits = result["bits"]
adc_code = result["adc_code"]
adc_reconstructed = result["adc_reconstructed"]
ideal_code = result["ideal_code"]
ideal_reconstructed = result["ideal_reconstructed"]
bubble = result["bubble"]
metrics = result["metrics"]
polarity = result["polarity"]


# ============================================================
# PRINT FINAL RESULTS
# ============================================================

codes_used = np.unique(adc_code)
not_sampled_codes = sorted(set(range(N_CODES)) - set(codes_used))

print()
print("===================================")
print("DYNAMIC RESULTS")
print("===================================")
print(f"Sampling phase       = {SAMPLE_PHASE:.3f} Ts")
print(f"ADC polarity         = {polarity}")
print(f"fin                  = {FIN/1e6:.3f} MHz")
print(f"FS ADC               = {FS/1e6:.3f} MHz")
print(f"N FFT                = {metrics['N']}")
print(f"Fundamental bin      = {metrics['fund_bin']}")
print(f"Fundamental freq     = {metrics['fund_freq']/1e6:.6f} MHz")
print(f"SINAD                = {metrics['SINAD']:.2f} dB")
print(f"ENOB                 = {metrics['ENOB']:.2f} bits")
print(f"SNR                  = {metrics['SNR']:.2f} dB")
print(f"THD                  = {metrics['THD']:.2f} dB")
print(f"SFDR                 = {metrics['SFDR']:.2f} dB")
print(f"Largest spur         = {metrics['spur_freq']/1e6:.6f} MHz")
print(f"Minimum code         = {np.min(adc_code)}")
print(f"Maximum code         = {np.max(adc_code)}")
print(f"Used codes at selected sampling phase = {codes_used}")
print(f"Codes not sampled at selected phase   = {not_sampled_codes}")

if raw_code_presence is not None:
    print(f"Observed codes in raw waveform        = {raw_code_presence['observed_codes']}")
    print(f"Missing codes in raw waveform         = {raw_code_presence['missing_codes']}")

    skipped_but_present = sorted(
        set(not_sampled_codes) - set(raw_code_presence["missing_codes"])
    )

    if len(skipped_but_present) > 0:
        print()
        print("NOTE:")
        print(f"Codes {skipped_but_present} were not captured at the selected sampling phase,")
        print("but they are present in the raw waveform.")
        print("Therefore, they are skipped sampled codes, not true missing codes.")


print()
print("===================================")
print("BUBBLE ERRORS")
print("===================================")
print(f"Analyzed samples        = {bubble['n_samples']}")
print(f"Samples with bubble     = {bubble['n_bubble_samples']}")
print(f"Bubble rate             = {bubble['bubble_rate_percent']:.4f} %")
print(f"Mean bubble strength    = {bubble['mean_bubble_strength']:.4f}")
print(f"Maximum bubble strength = {bubble['max_bubble_strength']}")


# ============================================================
# SINE / SLEW RATE RESULTS
# ============================================================

VIN_AMPLITUDE_EST = (np.max(vin_sampled) - np.min(vin_sampled)) / 2
VIN_OFFSET_EST = (np.max(vin_sampled) + np.min(vin_sampled)) / 2

SR_THEORETICAL = 2 * np.pi * FIN * VIN_AMPLITUDE_EST
SR_THEORETICAL_V_PER_US = SR_THEORETICAL / 1e6

print()
print("===================================")
print("SINE / SLEW RATE")
print("===================================")
print(f"Estimated amplitude  = {VIN_AMPLITUDE_EST:.3f} V")
print(f"Estimated offset     = {VIN_OFFSET_EST:.3f} V")
print(f"Estimated max SR     = {SR_THEORETICAL_V_PER_US:.3f} V/us")


# ============================================================
# EXPORT DATA
# ============================================================

if EXPORT_CSV:

    time_data = pd.DataFrame({
        "time_s": t_sample,
        "time_us": t_sample * 1e6,
        "vin_sampled_v": vin_sampled,
        "adc_code_measured": adc_code,
        "adc_reconstructed_measured_v": adc_reconstructed,
        "ideal_code_reference": ideal_code,
        "ideal_reconstructed_reference_v": ideal_reconstructed,
        "bubble_flag": bubble["bubble_flags"].astype(int),
        "n_transitions": bubble["n_transitions"],
        "bubble_strength": bubble["bubble_strength"]
    })

    for k in range(bits.shape[1]):
        time_data[f"bit_out{k}"] = bits[:, k]
        time_data[f"out{k}_sampled_v"] = outs_sampled[:, k]

    time_data.to_csv("adc_time_data.csv", index=False)

    fft_data = pd.DataFrame({
        "frequency_Hz": metrics["freq"],
        "frequency_MHz": metrics["freq"] / 1e6,
        "power_linear": metrics["P"],
        "power_dB": 10*np.log10(metrics["P"] + 1e-30)
    })

    fft_data.to_csv("adc_fft_data.csv", index=False)

    summary_data = pd.DataFrame({
        "metric": [
            "FIN_Hz",
            "FS_Hz",
            "sampling_phase_Ts",
            "N_FFT",
            "fundamental_bin",
            "fundamental_freq_Hz",
            "SINAD_dB",
            "ENOB_bits",
            "SNR_dB",
            "THD_dB",
            "SFDR_dB",
            "largest_spur_freq_Hz",
            "min_code",
            "max_code",
            "bubble_rate_percent",
            "sine_amplitude_est_v",
            "sine_offset_est_v",
            "slew_rate_v_per_us"
        ],
        "value": [
            FIN,
            FS,
            SAMPLE_PHASE,
            metrics["N"],
            metrics["fund_bin"],
            metrics["fund_freq"],
            metrics["SINAD"],
            metrics["ENOB"],
            metrics["SNR"],
            metrics["THD"],
            metrics["SFDR"],
            metrics["spur_freq"],
            np.min(adc_code),
            np.max(adc_code),
            bubble["bubble_rate_percent"],
            VIN_AMPLITUDE_EST,
            VIN_OFFSET_EST,
            SR_THEORETICAL_V_PER_US
        ]
    })

    summary_data.to_csv("adc_summary_results.csv", index=False)

    if DO_PHASE_SWEEP:
        phase_sweep_data = pd.DataFrame(phase_results)
        phase_sweep_data.to_csv("adc_phase_sweep_results.csv", index=False)

    if raw_code_presence is not None:

        raw_code_presence_data = pd.DataFrame({
            "code": np.arange(N_CODES),
            "raw_count": raw_code_presence["code_counts"],
            "raw_accumulated_time_s": raw_code_presence["code_time"],
            "raw_accumulated_time_ns": raw_code_presence["code_time"] * 1e9,
            "observed_in_raw_waveform": raw_code_presence["code_time"] > MIN_CODE_OBS_TIME
        })

        raw_code_presence_data.to_csv("adc_raw_code_presence.csv", index=False)

        raw_time_data = pd.DataFrame({
            "time_s": t_raw,
            "time_us": t_raw * 1e6,
            "vin_raw_v": vin_raw,
            "adc_code_raw_observed": raw_code_presence["code_raw"]
        })

        for k in range(raw_code_presence["bits_raw"].shape[1]):
            raw_time_data[f"bit_raw_out{k}"] = raw_code_presence["bits_raw"][:, k]

        raw_time_data.to_csv("adc_raw_time_code_data.csv", index=False)

    print()
    print("===================================")
    print("CSV FILES EXPORTED")
    print("===================================")
    print("adc_time_data.csv")
    print("adc_fft_data.csv")
    print("adc_summary_results.csv")

    if DO_PHASE_SWEEP:
        print("adc_phase_sweep_results.csv")

    if raw_code_presence is not None:
        print("adc_raw_code_presence.csv")
        print("adc_raw_time_code_data.csv")


# ============================================================
# PLOTS — TIME DOMAIN
# ============================================================

t_us = t_sample * 1e6
t0_us = t_us[0]
t1_us = t0_us + PLOT_WINDOW_US

plot_mask = (t_us >= t0_us) & (t_us <= t1_us)

plt.figure(figsize=(12,4))
plt.plot(
    t_us[plot_mask],
    vin_sampled[plot_mask],
    label="Vin",
    linewidth=1.5
)
plt.step(
    t_us[plot_mask],
    adc_reconstructed[plot_mask],
    where="post",
    label="ADC reconstructed — sampled",
    alpha=0.75
)
plt.step(
    t_us[plot_mask],
    ideal_reconstructed[plot_mask],
    where="post",
    label="ADC reconstructed — ideal reference",
    alpha=0.65,
    linestyle="--"
)
plt.title("Vin vs reconstructed ADC output — synchronous samples")
plt.xlabel("Time [µs]")
plt.ylabel("Voltage [V]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,4))
plt.step(
    t_us[plot_mask],
    adc_code[plot_mask],
    where="post"
)
plt.plot(
    t_us[plot_mask],
    adc_code[plot_mask],
    marker=".",
    linestyle="None",
    markersize=3
)
plt.title("ADC code at selected sampling phase")
plt.xlabel("Time [µs]")
plt.ylabel("Code")
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,4))
plt.step(
    t_us[plot_mask],
    bubble["bubble_flags"][plot_mask].astype(int),
    where="post"
)
plt.plot(
    t_us[plot_mask],
    bubble["bubble_flags"][plot_mask].astype(int),
    marker=".",
    linestyle="None",
    markersize=3
)
plt.title("Bubble errors — selected sampling phase")
plt.xlabel("Time [µs]")
plt.ylabel("Bubble error")
plt.yticks([0, 1], ["No bubble", "Bubble"])
plt.grid(True)
plt.tight_layout()
plt.show()


# ============================================================
# PLOTS — FFT
# ============================================================

plt.figure(figsize=(12,5))
plt.plot(metrics["freq"] / 1e6, 10*np.log10(metrics["P"] + 1e-30))
plt.axvline(FIN/1e6, color="red", linestyle="--", label="Fundamental")
plt.axvline(metrics["spur_freq"]/1e6, color="orange", linestyle="--", label="Largest spur")
plt.title("FFT of reconstructed ADC output")
plt.xlabel("Frequency [MHz]")
plt.ylabel("Power [dB]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# PLOTS — SAMPLED CODE HISTOGRAM
# ============================================================

plt.figure(figsize=(8,4))
plt.hist(
    adc_code,
    bins=np.arange(-0.5, N_COMPARATORS + 1.5, 1),
    edgecolor="black"
)
plt.title("ADC code histogram — selected sampling phase")
plt.xlabel("Code")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()


# ============================================================
# PLOTS — RAW WAVEFORM CODE PRESENCE
# ============================================================

if raw_code_presence is not None:

    plt.figure(figsize=(8,4))
    plt.bar(
        np.arange(N_CODES),
        raw_code_presence["code_time"] * 1e9,
        edgecolor="black"
    )
    plt.title("Code presence in raw waveform")
    plt.xlabel("ADC code")
    plt.ylabel("Accumulated time [ns]")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12,4))

    raw_t_us = t_raw * 1e6
    raw_plot_mask = (
        (raw_t_us >= t_us[0])
        & (raw_t_us <= t_us[0] + PLOT_WINDOW_US)
    )

    plt.step(
        raw_t_us[raw_plot_mask],
        raw_code_presence["code_raw"][raw_plot_mask],
        where="post"
    )

    plt.title("ADC code observed in raw waveform")
    plt.xlabel("Time [µs]")
    plt.ylabel("ADC code")
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# ============================================================
# PLOTS — COMPARATOR OUTPUTS AT SELECTED SAMPLING PHASE
# ============================================================

plt.figure(figsize=(12,6))
for k in range(bits.shape[1]):
    plt.step(
        t_us[plot_mask],
        bits[plot_mask, k] + 1.2*k,
        where="post",
        label=f"out{k}"
    )

plt.title("Comparator digital outputs — selected sampling phase")
plt.xlabel("Time [µs]")
plt.ylabel("Comparator index offset")
plt.grid(True)
plt.tight_layout()
plt.show()


# ============================================================
# OPTIONAL DOWNLOAD OF CSV FILES
# ============================================================

if EXPORT_CSV:
    files.download("adc_time_data.csv")
    files.download("adc_fft_data.csv")
    files.download("adc_summary_results.csv")

    if DO_PHASE_SWEEP:
        files.download("adc_phase_sweep_results.csv")

    if raw_code_presence is not None:
        files.download("adc_raw_code_presence.csv")
        files.download("adc_raw_time_code_data.csv")
```
